## Path generation using linear algebra

When using Monte Carlo simulation to price a vanilla European call option in the Black-Scholes world, only one draw from the standard normal distribution is needed per simulation, because the underlying stock's price is only needed *at expiration* in order to determine the payoff. If $N(0,1)$ denotes a standard normal random variable, then the geometric Brownian model for the stock price $S_T$ at time $T$ (the expiration date of the option, i.e., $T$ years from now) is $S_T = S_0 e^{\left( r - \frac{\sigma^2}{2} \right)T + \sigma \sqrt{T} N(0,1)}$, where we have assumed, for option pricing purposes, that the stock grows at the risk-free interest rate $r$ (here $\sigma$ is, as usual, the volatility). The corresponding value of a call option struck at $K$ is then $C = e^{-rT}(\mathrm{max}(S_T - K, 0))$, and we can approximate the true value of $C$ by taking many independent $N(0,1)$ draws and then averaging the corresponding values of $C$. Note that here none of the values of $S$ between times $0$ and $T$ were needed in the formula.

However, there are other cases in which we need to know the intermediate values of $S$, not just the endpoints, to carry out a Monte Carlo simulation. One case, discussed in the lectures (and featured in the subsequent Mini Project 4) is simulation of a *dynamically-hedged* portfolio, where stock is bought or sold at several intermediate times in order to reduce the risk of an option position. Another case, discussed below, is that of a *path-dependent* option, where in order even to know the final payoff of the option one must know the value of the underlying stock at multiple specified times. One example of a path-dependent option is an **average price option (APO)**, commonly known as an **Asian option** because they were introduced in Tokyo in the 1980s. Asian options are typically European-style (no early exercise features), but there are several variants (geometric vs. arithmetic average, average price vs. average strike, etc.) For concreteness we will focus exclusively on an arithmetic average-price Asian call option. 

The payoff of a vanilla call option at expiration $T$ is $\mathrm{max}(S_T - K, 0)$. In contrast, the payoff of an arithmetic average-price call option (which we will just call an "Asian option" for the remainder of this mini-project) is $\mathrm{max}(\frac{1}{n} \sum_{i=1}^n S_{t_i} - K, 0)$, where $0 = t_0 < t_1 < \cdots < t_n = T$ is a pre-specified sequence of times ending with the expiration time $T$. That is, when calculating the final payoff, the final stock price is replaced by the **average** of the stock prices at times $t_1, t_2, \ldots, t_n$ (the "look-at" times/dates). One example we will have in mind throughout this mini-project is an Asian option expiring in one year with weekly look-at dates, so the payoff depends on the average of the spot price at $52$ (equally-spaced) times.

One reason you may wish to purchase an Asian option is to "smooth out" exposure to something volatile over a long period. For example, if you are going to receive payments in a foreign currency weekly or monthly for a year, the value of your cashflows fluctuates with the exchange rate. Holding an Asian option on the exchange rate expiring in a year allows you to treat these cashflows, at least retroactively, as the same, and the exchange rate averaged over a year may be more predictable. Asian options are also popular among investors with exposure to the price of oil or other energy markets. The CME Group explains: *"APOs are optimal hedging tools for markets that can be prone to price shocks and spikes. Also, they can be ideal for markets with infrequent trades-- as these options’ payoff is less sensitive to extreme market fluctuations especially toward the option expiration."*

A well-known mathematical result, the **Arithmetic-Geometric Mean (AM-GM) Inequality** (beloved of math contest participants and problem setters), states that if $a_1, \ldots, a_n \geq 0$ are real numbers, then $(a_1 a_2 \cdots a_n)^{1/n} \leq \frac{a_1 + \cdots + a_n}{n}$. The number on the left is the *geometric mean* of $a_1, \ldots, a_n$, and this sort of average is often used instead of the arithmetic mean to define the payoff of an Asian option. Because of this inequality, it is clear that we should expect geometric-average Asian options to be cheaper than the arithmetic-average ones we are considering in this project. Both types of Asian options should be cheaper than a vanilla option, because the averaging reduces volatility over the life of the option.

In order to run a Monte Carlo simulation that returns the approximate price of an Asian option, we need to generate a *stock price path*: a sequence $(S_0, S_{t_1}, S_{t_2}, \ldots, S_{t_n})$ of spot prices following geometric Brownian motion with drift equal to the risk-free rate (because we are in a Black-Scholes world). For $i = 1, 2, \ldots, n$, the random variable $S_{t_i}$ is distributed as $S_{t_i} = S_0 e^{\left( r - \frac{\sigma^2}{2} \right)t_i + \sigma \sqrt{t_i} X_i}$, where $X_i$ is a standard normal variable. However, the random variables $X_1, \ldots, X_n$ are *not* independent, because the price of $S$ at $t_1$ affects the distribution of possible prices at $t_2$ and so on. More formally, the random variables $\sqrt{t_i} X_i$ are $n$ of the uncountably many constituents of a **Wiener process**, a continuous-time stochastic process $(W_t)_{t \geq 0}$ where $W_0 = 0$, $W_t - W_s$ (for $t \geq s$) is always normally distributed with mean $0$ and variance $t - s$, and $W$ has *independent increments* in the sense that if $t \geq s$, then $W_s$ and $W_t - W_s$ are independent. A Wiener process has continuous sample paths, and our $X_1, \ldots, X_n$ must be taken from a single sample path in order for our model of a stock price path to make sense. 

There are two different approaches we can take to the problem of drawing a multi-step stock price path. The first is **incremental generation**, where we first produce $S_{t_1}$ (the same way we generated the terminal spot price for the vanilla option), then produce $S_{t_2}$ taking into account the already-known value of $S_{t_1}$, and so on. If we use this method, we have

$S_{t_1} = S_0 e^{\left( r - \frac{\sigma^2}{2} \right)t_1 + \sigma \sqrt{t_1} Z_1}$,

then ($S_{t_1}$ now being known) $S_{t_2} = S_{t_1} e^{\left( r - \frac{\sigma^2}{2} \right)(t_2 - t_1) + \sigma \sqrt{t_2 - t_1}Z_2}$,

and in general, $S_{t_{i+1}} = S_{t_i} e^{\left( r - \frac{\sigma^2}{2} \right)(t_{i+1} - t_i) + \sigma \sqrt{t_{i+1}-t_i} Z_{i+1}}$,

where $Z_1, \ldots, Z_n$ are *independent* $N(0,1)$ random variables: the dependence of $S_{t_{i+1}}$ on $S_{t_i}$ is represented by the factor of $S_{t_i}$ in the formula. As we will see below, generation of such an incremental path can be done in a vectorized manner using `numpy`, avoiding the need to define the spot prices recursively using a loop. This approach was followed in the lectures on dynamic hedging and so we will more-or-less copy that code here.

In the other approach, we will generate the entire path $(S_0, S_{t_1}, S_{t_2}, \ldots, S_{t_n})$ at once. We need to produce random variables $X_1, \ldots, X_n$ where $X_i$ plays the role of $\sqrt{t_1} Z_1 + \sqrt{t_2 - t_1} Z_2 + \cdots + \sqrt{t_i - t_{i-1}} Z_i$. Recall that the Wiener process $(W_t)$ whose sample paths give us simulated stock price paths has the independent-increments property. If $s \leq t$, what is the covariance of $W_s$ and $W_t$? By definition, this is $\mathrm{Cov}(W_s, W_t) = E(W_sW_t) - E(W_s)E(W_t)$. Each $W_r$ is an $N(0,r)$ random variable, so the second term is $0 \cdot 0 = 0$, and the first can be rewritten $E(W_s(W_t-W_s) + W_s^2) = E(W_s(W_t-W_s)) + E(W_s^2)$. By the independence of increments, $E(W_s(W_t-W_s)) = E(W_s)E(W_t-W_s)$, which is $0$ since both $W_s$ and $W_t - W_s$ are normal random variables with mean $0$. Finally, $E(W_s^2) = \mathrm{Var}(W_s) = s$. We conclude that $\mathrm{Cov}(W_s, W_t) = s = \mathrm{min}\{s,t\}$. This means that the covariance matrix $C$ of our desired random variables $X_i$ (which, when generated, will be plugged into the formulas $S_{t_i} = S_0 e^{\left( r - \frac{\sigma^2}{2} \right)t_i + \sigma X_i}$ to generate our stock price path) has $(i,j)$-entry $C_{ij} = \mathrm{Cov}(X_i, X_j) = \mathrm{min}\{t_i, t_j\}$. 

Now recall that the sum of independent normal random variables is again normally distributed, and the means and variances of the summands simply add. More generally, we can take linear combinations of normally distributed variables to produce new ones. If $Z = \begin{bmatrix} Z_1 & \cdots & Z_n \end{bmatrix}^T$ is a vector of independent normal random variables, consider the matrix product $X = AZ$ where $A = (a_{ij})$ is an $n \times n$ matrix of real numbers. If we write $X = \begin{bmatrix} X_1 & \cdots & X_n \end{bmatrix}^T$, then each $X_i$ is a linear combination $a_{i1}Z_1 + \cdots + a_{in}Z_n$ of the $Z_j$s, and therefore is a normally distributed random variable. However, the $X_i$ are not independent. We can calculate their covariance matrix as follows: for each $i$ and $j$,

$\mathrm{Cov}(X_i, X_j) = \mathrm{Cov}\left( \sum_{k=1}^n a_{ik}Z_k, \sum_{l=1}^n a_{jl}Z_l \right) = \sum_{k,l=1}^n a_{ik}a_{jl} \mathrm{Cov}(Z_k, Z_l) = \sum_{k=1}^n a_{ik}a_{jk},$

which is just the dot product of the $i$th and $j$th rows of $A$, or in other words, the $(i,j)$-entry of $AA^T$. (We have used here the fact that the $Z_i$ are independent, so if $k \neq l$, the covariance of $Z_k$ and $Z_l$ is zero.) This means that $AA^T$ *is* the covariance matrix of the $X_i$. Applying this reasoning in reverse, if we would like to synthesize normally distributed random variables $X_i$ with a specific desired (positive semidefinite) covariance matrix $C$ given independent normal random variables $Z_i$, all we need to do is find a matrix $A$ such that $AA^T = C$ (a so-called **pseudo square root** of $A$), and then $X = AZ$ does the job where $X$ and $Z$ are the vectors of the $X_i$ and $Z_i$. 

To solve the problem of generating a stock price path, we therefore would like to find a pseudo square root $A$ of the matrix $C = (\mathrm{min}\{t_i, t_j\})_{i,j}$. When performing incremental path generation, the transformation from independent $N(0,1)$ random variables $Z_i$ to the correlated normal random variables $X_i$ was provided by the formulas $X_i = \sqrt{t_1} Z_1 + \sqrt{t_2 - t_1} Z_2 + \cdots + \sqrt{t_i - t_{i-1}} Z_i$. The coefficients appearing in front of the $Z_i$ would then be the entries of $A$, and since each $X_i$ depends only on the first $i$ $Z$s, the resulting matrix $A$ would be lower triangular. A factorization of a symmetric, positive-semidefinite matrix $C$ as $C = AA^T$ where $A$ is lower triangular is known as a **Cholesky decomposition**, and (if conditions are imposed on the order of the diagonal entries of $A$) is unique. So if we use a Cholesky decomposition of $C = (\mathrm{min}\{t_i, t_j\})_{i,j}$ to generate the stock price path, the result is equivalent to the incremental generation method, meaning that we have done nothing new.

However, there are other choices of pseudo square roots, leading to alternative, non-incremental methods for generating the path that have their own advantages and disadvantages. We describe an alternative method based on spectral decomposition (**orthogonal diagonalization**). By the spectral theorem, every symmetric $n \times n$ real matrix $C$ can be written as $PDP^T$ where $D$ is a diagonal matrix (whose diagonal entries are the eigenvalues $\lambda_i$ of $C$) and $P$ is an orthogonal matrix (whose columns $\vec{v}_i$ are eigenvectors for $C$ and form an orthonormal basis for $\mathbb{R}^n$, appearing in the same order as the eigenvalues do in $D$). Here the superscript $T$ denotes the *transpose* operation, which reverses the roles of rows and columns (for orthogonal matrices, the transpose and inverse coincide). If $C$ is positive-semidefinite, its eigenvalues are all nonnegative and so we can form the matrix $D^{1/2}$ whose diagonal entries are the square roots $\sqrt{\lambda_i}$ of the eigenvalues of $C$. Then if we set $A = PD^{1/2}$, we have

$AA^T = (PD^{1/2})(PD^{1/2})^T = PD^{1/2}(D^{1/2})^TP^T = PD^{1/2}D^{1/2}P^T = PDP^T = C$,

since diagonal matrices are symmetric and the product of two diagonal matrices is another diagonal matrix where the diagonals have been multiplied component-wise. That is, $PD^{1/2}$ is also a pseudo square root of $C$. There are different choices one can make when constructing the matrices $P$ and $D$, but below (for reasons to be elaborated upon) we will always assume that the diagonal entries of $D$ are the eigenvalues of $C$ *in decreasing order* (usefully for us, the `eig` method in `numpy.linalg` returns the eigenvalues in this order and the corresponding eigenvectors the same way).

Using this choice of pseudo square root and "unwinding" the definition of matrix multiplication, the transformation $Z \mapsto X = AZ$ from the vector $Z$ of independent $N(0,1)$ random variables $Z_i$ to the vector $X$ of correlated normal random variables $X_i$ that we will use to generate the path is given by $X = \sum_{j=1}^n \sqrt{\lambda_j} Z_j \vec{v}_j$.

We will compare our Monte Carlo prices for an Asian option (computed in three different ways: incremental path generation, orthogonal diagonalization of the covariance matrix, and a variant on the orthogonal diagonalization method to be described below) with the Black-Scholes value of a vanilla call option with the same parameters (strike, volatility, etc.) As stated above, we expect the price of the Asian option to be less than that of a vanilla, providing a safety check on our answers.

```python
# import packages
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def bs_call(S_0, K, sigma, T = 1, r = 0):
    """
    Description:
    
    Computes the Black-Scholes value of a European call option.
    
    Parameters:
        S_0: Current asset price
        K: Strike price
        sigma: Yearly standard deviation of log-returns (volatility)
        T: Time to expiration (in years)
        r: Risk-free interest rate
    
    Returns:
        Call option price
    """
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d_2 = d_1 - sigma*np.sqrt(T)
    return S_0*norm.cdf(d_1)-K*np.exp(-r*T)*norm.cdf(d_2)

print(bs_call(100, 110, 0.25, 1, 0.03))
```

First, we define a Python function `asian_monte` that uses incremental path generation to price an Asian option by Monte Carlo. Because of `numpy`'s vectorization capabilities, the incremental nature of the process is implicit. 

This pricer assumes that the "look-at" times $t_1, t_2, \ldots$ are equally spaced between time zero and expiration. (Again, the example we will use below is a one-year option with weekly look-at dates.) If the Asian option is defined using non-equally spaced averaging times, the `times` array would need to be modified.

```python
# Monte Carlo pricer for an Asian option

def asian_monte(S_0, K, sigma, T, r, n_steps, n_sims):
    """
    Description:
    
    Computes the price for an arithmetic-average-price Asian call option (in a Black-Scholes world) by Monte Carlo simulation.
    
    Parameters:
        S_0: Current asset price (float)
        K: Strike price (float)
        sigma: Yearly standard deviation of log-returns, a.k.a. volatility (float)
        T: Time to expiration in years (float)
        r: Risk-free interest rate (float)
        n_steps: Number of equally-spaced look-at times. The Asian option's payoff depends on the average of the spot price at these times. (int)
        n_sims: Number of simulations to use in the Monte Carlo average (int)
    
    Returns:
        Simulated Asian call option price (float)
    """
    dt = T/n_steps 
    times = np.linspace(dt, T, n_steps) # array of look-at times t_1, t_2, ..., t_n = T (does not include time zero)

    # simulate n_sims stock price paths, each of length n_steps
    uncorrelated_steps = np.random.normal(0, 1, size=(n_sims, n_steps))
    log_returns = (r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * uncorrelated_steps
    log_paths = np.cumsum(log_returns, axis=1)
    S_paths = S_0 * np.exp(log_paths) # each row of the array represents a distinct simulation
    
    asian_payoffs = np.maximum(np.mean(S_paths, axis=1)-K, 0) # average stock price along rows, then compare with strike K to see if option pays off
    return np.exp(-r*T)*np.mean(asian_payoffs) # discount average payoff to time zero to get option price

print(asian_monte(100, 110, 0.25, 1, 0.03, 52, 10000))
```

As predicted, an Asian call option with the same spot price, strike price, volatility, time to expiration and risk-free rate as a corresponding vanilla call is significantly cheaper than the vanilla, because the averaging effect reduces the risk. Here we see that the Asian option with weekly look-at dates (hence, whose payoff depends on the average of $52$ values of spot) is approximately \\$2.71, compared with a Black-Scholes price of \\$7.26 for the vanilla call. Let's see what happens to the Asian option when we reduce the frequency of the look-at dates, taking the average monthly or even quarterly instead of weekly.

```python
print(asian_monte(100, 110, 0.25, 1, 0.03, 52, 10000)) # weekly look-at dates
print(asian_monte(100, 110, 0.25, 1, 0.03, 12, 10000)) # monthly look-at dates
print(asian_monte(100, 110, 0.25, 1, 0.03, 4, 10000)) # quarterly look-at dates
print(asian_monte(100, 110, 0.25, 1, 0.03, 1, 10000)) # yearly look-at dates; with 1 year to expiration, this should just behave like a vanilla call
print(bs_call(100, 110, 0.25, 1, 0.03)) # vanilla call price for comparison
```

The fewer dates over which we average, the riskier and hence more expensive the option becomes.

Next, we build another Monte Carlo pricer using the pseudo square root $A = PD^{1/2}$ from the orthogonal diagonalization of the covariance matrix. Again we assume that the look-at dates are equally spaced, so if $n$ is the number of look-at dates and $T$ the time to expiration, the $(i,j)$-entry of the $n \times n$ covariance matrix $C$ is $\mathrm{min}\{iT/n,jT/n\}$. The function `la_asian_monte` begins by setting up the matrix $C$ and finding its eigenvalues and eigenvectors (that is, finding the $D$ and $P$ in an orthogonal diagonalization). The uniform random draws `uncorrelated_steps`, as well as the final payoff calculation, averaging, and discounting, are copied from `asian_monte`, but the construction of the simulated paths is now done by matrix multiplication with the pseudo square root $A = PD^{1/2}$ of $C$. 

The use of transposes in the code may be confusing, so it is perhaps best to explain it here rather than in code comments. As described earlier, we are transforming a vector $Z$ of independent $N(0,1)$ draws into a vector $X$ of normal random variables with the desired correlations by multiplying by the pseudo square root: $X = AZ$. The vectors $X$ and $Z$ are $n \times 1$ matrices, or column vectors, where $n$ is the number of look-at dates (steps in each simulated path, or `n_steps` in the code), and $A$ is an $n \times n$ matrix. However, as we are interested in doing this simulation many times and not just once, we begin instead with a matrix $\mathbf{Z}$ of independent $N(0,1)$ draws, of size `n_sims` by `n_steps`. To be consistent with our earlier function `asian_monte` as well as the Monte Carlo examples from the boot camp Jupyter notebooks and lectures, each *row* of $\mathbf{Z}$ should represent a simulation, and each *column* a time step. That is, the rows of $\mathbf{Z}$ should each be thought of as a separate $Z$ in the earlier notation. The outcome of the multiplication, $\mathbf{X}$, should also have rows (not columns) that correspond to separate simulations. By using transposes, we can do all the separate "$X = AZ$" calculations in a single matrix multiplication while simultaneously switching the separate simulations to be represented by rows and not columns:

`correlated_steps` = $\mathbf{X} = (\mathbf{X}^T)^T = (A\mathbf{Z}^T)^T = \mathbf{Z}A^T$, the matrix product of $\mathbf{Z}$ = `uncorrelated_steps` and $A^T$.

(Here, we have used the fact that for all pairs of matrices $M$ and $N$ such that $MN$ exists, $(MN)^T = N^TM^T$.)

```python
# Monte Carlo pricer for an Asian option, using orthogonal diagonalization to generate the simulated stock price paths

def la_asian_monte(S_0, K, sigma, T, r, n_steps, n_sims): # "la": linear algebra
    """
    Description:
    
    Computes the price for an arithmetic-average-price Asian call option (in a Black-Scholes world) by Monte Carlo simulation.
    
    Parameters:
        S_0: Current asset price (float)
        K: Strike price (float)
        sigma: Yearly standard deviation of log-returns, a.k.a. volatility (float)
        T: Time to expiration in years (float)
        r: Risk-free interest rate (float)
        n_steps: Number of equally-spaced look-at times. The Asian option's payoff depends on the average of the spot price at these times. (int)
        n_sims: Number of simulations to use in the Monte Carlo average (int)
    
    Returns:
        Simulated Asian call option price (float)
    """

    times = np.linspace(T/n_steps,T,n_steps)
    covariance_matrix = np.empty((n_steps,n_steps))
    for i in range(n_steps):
        for j in range(n_steps):
            covariance_matrix[i][j] = min(times[i], times[j]) # (i,j)-entry is now the minimum of iT/n and jT/n

    cv_eigeninfo = np.linalg.eig(covariance_matrix) # eig returns array of eigenvalues, cv_eigeninfo[0], then array of eigenvectors, cv_eigeninfo[1]
    cv_eigenvectors = cv_eigeninfo[1]
    cv_diag = np.zeros((n_steps, n_steps)) # cv_eigeninfo[0] is a 1-D array, but we want 2-D matrices for D and D^{1/2}
    for i in range(n_steps):
        cv_diag[i][i] = np.sqrt(cv_eigeninfo[0][i]) # cv_diag is D^{1/2}, with the square root of the i-th largest eigenvalue in the (i,i)-position

    pseudo_root = np.matmul(cv_eigenvectors, cv_diag) # this is A = PD^{1/2}, so the covariance matrix C is AA^T

    uncorrelated_steps = np.random.normal(0,1,size=(n_sims, n_steps))
    correlated_steps = np.matmul(uncorrelated_steps, np.transpose(pseudo_root)) # explained in block above: recall that (MN)^T = N^TM^T
    
    times_grid = np.broadcast_to(times,(n_sims,n_steps)) # make an array of times, with each row representing a new simulation (all rows identical)
    exponents = (r - 0.5*(sigma**2))*times_grid + sigma*correlated_steps 
    stock_price_paths = S_0*np.exp(exponents) # each row of the array represents a distinct simulation
    
    asian_option_payoffs = np.maximum(np.mean(stock_price_paths, axis=1)-K,0)
    return np.exp(-r*T)*np.mean(asian_option_payoffs)

print(la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 10000))
```

In our example above, the call to `la_asian_monte` required the diagonalization of ten thousand $52 \times 52$ matrices. We will compare the run time of `la_asian_monte` and a variant of it with the run time of the incremental path method below to see if the size of these matrices is a significant computational bottleneck.

Why introduce `la_asian_monte` at all when incremental path generation works just fine? As we will see below, `la_asian_monte` does not actually offer us a significant advantage in computing time or accuracy. The real reason to care about this diagonalization-based approach is *dimensionality reduction*. Loosely put, it takes many more (exponentially more) $n$-tuples of $N(0,1)$ draws to "cover" an $n$-dimensional hypercube with side length $k$ than it takes individual $N(0,1)$ draws to cover an interval of length $k$. Monte Carlo simulation of an Asian option expiring in one year with weekly look-at dates is a $52$-dimensional problem, in that we need $52$-tuples of stock prices that form a representative sample of all such paths in order to accurately simulate the payoff distribution and hence decide on a price. The rate of growth of the time (number of simulations) needed to get an accurate price as the dimension grows is a significant computational problem and is sometimes referred to as the *curse of dimensionality*.

Now recall that the transformation from the vector $Z$ of independent $N(0,1)$ random variables $Z_i$ to the vector $X$ of correlated normal random variables $X_i$ used in `la_asian_monte` was $X = \sum_{j=1}^n \sqrt{\lambda_j} Z_j \vec{v}_j$. The eigenvalues $\lambda_j$, and hence their square roots, are listed in *decreasing order*; for large matrices, the eigenvalues at the tail end of the sum can be quite small indeed. This means that the terms at the beginning of the sum contribute more, and we may hope that a truncated sum, $\sum_{j=1}^k \sqrt{\lambda_j} Z_j \vec{v}_j$ for some choice of $k < n$, approximates the full sum well. (One might object that the components of the vectors $\vec{v_j}$ appearing at the tail end of the sum could be large enough to compensate for the $\lambda$s being smaller, but since the $\vec{v}_j$ form an orthonormal set, their Pythagorean lengths are $1$ and therefore none of their entries can exceed $1$ in absolute value.)

Below we define a variant of `la_asian_monte` that takes an additional parameter, the `truncation_index` which indicates how many terms we are to keep. It is important to emphasize here that we still need to compute the eigenvectors of the full-size matrix. For example, if $n = 52$ and we take $k = 10$ for our truncation index, the sum $\sum_{j=1}^k \sqrt{\lambda_j} Z_j \vec{v}_j$ has only ten terms but the vectors $\vec{v}_j$ still belong to $\mathbb{R}^{52}$. This means the matrices we need to diagonalize are still the same size, and so we should not expect each individual simulation to go that much faster. What we are doing instead is focusing our attention on a $10$-dimensional subspace of $\mathbb{R}^{52}$, namely the subspace spanned by the first ten $\vec{v}_i$, reducing the dimensionality of our simulation problem. We therefore expect that the approximate prices generated by our simulations should *converge more quickly*, meaning that we won't need as many simulations before we're confident we have the right price.

Besides the truncation, we make one other tweak to the code below. By truncating the sum used to define the $X_i$, we have decreased their variance. We therefore multiply by a scaling factor,

$\alpha = \sqrt{\frac{\sum_{i=1}^n \lambda_i}{\sum_{i=1}^k \lambda_i}} \geq 1$,

to keep the variance the same as it was in the non-truncated version of the simulation.

When selecting $k$, there is a trade-off between the increased speed of convergence and the decreased accuracy resulting from the missing terms of the sum. If we take $k$ to be too small (chop off too many terms), we run the risk that the price will converge quickly, but to a very wrong number! We will experiment below with different values of $k$ and different numbers of simulations.

What we are doing here is a form of **principal component analysis (PCA)**. Every matrix $M$ (of any size, say $m \times n$, not just square) has a **singular value decomposition (SVD)** $M = U \Sigma V^T$ where $U$ and $V$ are orthogonal matrices (of respective sizes $m \times m$ and $n \times n$) and $\Sigma$ is an $m \times n$ matrix with the square roots of the eigenvalues of $M^TM$ on its main diagonal and $0$s elsewhere. The columns of $V$ (rows of $V^T$) in such a decomposition are called the *principal components*, and the first few principal components represent the directions of greatest variance (keeping only the first few principal components is an attempt to preserve the basic shape of the data in a smaller-dimensional space). The SVD is not in general the same thing as a diagonalization, but in our special case, $C$ is symmetric and positive-semidefinite, so $C^TC = C^2$ which has the same eigenvectors as $C$. The square roots of the eigenvalues of $C^TC = C^2$ are just the original (nonnegative) eigenvalues of $C$ itself. So $C = PDP^T$ actually is an SVD of $C$ in this case, and the vectors $\vec{v}_i$ we calculated *are* the principal components.

```python
# Monte Carlo pricer for an Asian option, using orthogonal diagonalization to generate the simulated stock price paths,
# and keeping only some principal components to reduce dimensionality

def truncated_la_asian_monte(S_0, K, sigma, T, r, n_steps, n_sims, truncation_index):
    """
    Description:
    
    Computes the price for an arithmetic-average-price Asian call option (in a Black-Scholes world) by Monte Carlo simulation.
    
    Parameters:
        S_0: Current asset price (float)
        K: Strike price (float)
        sigma: Yearly standard deviation of log-returns, a.k.a. volatility (float)
        T: Time to expiration in years (float)
        r: Risk-free interest rate (float)
        n_steps: Number of equally-spaced look-at times. The Asian option's payoff depends on the average of the spot price at these times. (int)
        n_sims: Number of simulations to use in the Monte Carlo average (int)
        truncation_index: 
    
    Returns:
        Simulated Asian call option price (float)
    """

    times = np.linspace(T/n_steps,T,n_steps)
    covariance_matrix = np.empty((n_steps,n_steps))
    for i in range(n_steps):
        for j in range(n_steps):
            covariance_matrix[i][j] = min(times[i], times[j])

    cv_eigeninfo = np.linalg.eig(covariance_matrix) 
    cv_eigenvectors = cv_eigeninfo[1]
    cv_diag = np.zeros((n_steps, n_steps)) # so far, same as la_asian_monte
    
    # scaling factor to keep variance after truncating
    alpha = np.sqrt(np.sum(cv_eigeninfo[0])/np.sum(cv_eigeninfo[0][:truncation_index]))
    for i in range(n_steps):
        cv_diag[i][i] = np.sqrt(cv_eigeninfo[0][i]) # this is D^{1/2}

    truncated_pseudo_root = np.matmul(cv_eigenvectors[:n_steps, :truncation_index], cv_diag[:truncation_index, :truncation_index])
    # we use only the first k = truncation_index eigenvectors to transform the Zs into the Xs
    # P is truncated to size N x k (N = n_steps), D^{1/2} is truncated to size k x k (so PD^{1/2} is now N x k)

    uncorrelated_steps = np.random.normal(0,1,size=(truncation_index, n_sims))
    correlated_steps = np.transpose(alpha*np.matmul(truncated_pseudo_root, uncorrelated_steps))
    # original variance preserved after multiplying by alpha
    # PD^{1/2} is N x k, uncorrelated_steps is k x n_sims, so correlated_steps is n_sims x N
    # each row represents a simulation
    # now everything below works the same as in la_asian_monte
    
    times_grid = np.broadcast_to(times,(n_sims,n_steps))
    exponents = (r - 0.5*(sigma**2))*times_grid + sigma*correlated_steps 
    stock_price_paths = S_0*np.exp(exponents)
    asian_option_payoffs = np.maximum(np.mean(stock_price_paths, axis=1)-K,0)
    return np.exp(-r*T)*np.mean(asian_option_payoffs)

print(la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 10000))
```

Below, we use IPython's `%timeit` "magic function" to compare the run time of `asian_monte`, `la_asian_monte`, and `truncated_la_asian_monte`. In all three cases we use our example of a one-year option with weekly look-at dates. We look at the run time with $100000$ simulations compared with a million, and in the case of the truncated version, we compare the run times for truncation indices $5$, $10$, and $20$ (keeping only five, ten, or twenty of the $52$ components).

```python
%timeit print(asian_monte(100, 110, 0.25, 1, 0.03, 52, 100000))
%timeit print(asian_monte(100, 110, 0.25, 1, 0.03, 52, 1000000))

%timeit print(la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 100000))
%timeit print(la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 1000000))

%timeit print(truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 100000, 5))
%timeit print(truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 1000000, 5))

%timeit print(truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 100000, 10))
%timeit print(truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 1000000, 10))

%timeit print(truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 100000, 20))
%timeit print(truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 1000000, 20))
```

To summarize the results (focusing for now only on speed, not accuracy):

- `asian_monte` (incremental path generation) took 0.364 seconds to run a hundred thousand simulations and 3.48 seconds to run a million.

- `la_asian_monte` (orthogonal diagonalization, keeping all 52 principal components) took 0.409 seconds to run a hundred thousand simulations and 3.76 seconds to run a million, slightly slower than `asian_monte`.

- If only five principal components were used, `truncated_la_asian_monte` took 0.235 seconds to run a hundred thousand simulations and 2.18 seconds to run a million.

- If ten principal components were used, `truncated_la_asian_monte` took 0.258 seconds to run a hundred thousand simulations and 2.49 seconds to run a million.

- If twenty principal components were used, `truncated_la_asian_monte` took 0.297 seconds to run a hundred thousand simulations and 2.83 seconds to run a million.

So even though the matrices being diagonalized in the truncated version are the same size, the reduced number of multiplications needed afterward appears to save us time, somewhat counter to our expectations above. However, simulation *speed* was not the primary reason for truncating, but rather simulation *accuracy/convergence*. To understand this, we need to look at each method and run more and more simulations, keeping track of how long it takes for the simulated price to "settle down" to a limiting value. Specifically, we will run $2^k$ simulations using each method, as $k$ ranges from $10$ to $20$, and keep track of the absolute value of the difference between the price generated by a run of $2^k$ simulations and a run of $2^{k+1}$.

```python
asian_monte_sims = [asian_monte(100, 110, 0.25, 1, 0.03, 52, 2**k) for k in range(10,21)] 
# generates a list of approximate prices based on 2^10, 2^11, ..., 2^20 simulations

asian_monte_diff = [abs(asian_monte_sims[i] - asian_monte_sims[i-1]) for i in range(1,11)]
# generates a list of the absolute differences between successive powers of two

# now we do the same thing for all the other methods

la_asian_monte_sims = [la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 2**k) for k in range(10,21)]

la_asian_monte_diff = [abs(la_asian_monte_sims[i] - la_asian_monte_sims[i-1]) for i in range(1,11)]

t_five_asian_monte_sims = [truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 2**k, 5) for k in range(10,21)] # "t": truncated

t_five_asian_monte_diff = [abs(t_five_asian_monte_sims[i] - t_five_asian_monte_sims[i-1]) for i in range(1,11)]

t_ten_asian_monte_sims = [truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 2**k, 10) for k in range(10,21)]

t_ten_asian_monte_diff = [abs(t_ten_asian_monte_sims[i] - t_ten_asian_monte_sims[i-1]) for i in range(1,11)]

t_twenty_asian_monte_sims = [truncated_la_asian_monte(100, 110, 0.25, 1, 0.03, 52, 2**k, 20) for k in range(10,21)]

t_twenty_asian_monte_diff = [abs(t_twenty_asian_monte_sims[i] - t_twenty_asian_monte_sims[i-1]) for i in range(1,11)]

plt.figure(figsize = (12,8))

x = np.arange(10,20)
plt.plot(x, asian_monte_diff, label = "Incremental")
plt.plot(x, la_asian_monte_diff, label = "Orth Diag")
plt.plot(x, t_five_asian_monte_diff, label = "5 components")
plt.plot(x, t_ten_asian_monte_diff, label = "10 components")
plt.plot(x, t_twenty_asian_monte_diff, label = "20 components")
plt.title("Convergence of Asian option Monte Carlo with different path generation methods")
plt.xlabel("k")
plt.ylabel("Price difference between $2^k$ and $2^{k+1}$ simulations")
plt.legend()
plt.show()
```

The green graph, representing the truncated method where only five principal components are used, was the fastest to fall below a difference of $0.05$ (between a simulation count of one power of two and the next) and stay there.

Finally, we mention one last point about the pricing of Asian options. Suppose we hold an Asian call option with look-at times $t_1 < \cdots < t_n = T$ and strike price $K$, and consider our position at time $t_{n-1}$ (the second-to-last look-at time). At this time, the only "mystery" left is the price $S_{t_n} = S_T$ and whether its contribution to the average will send that average below $K$. The final payoff of the Asian option is $\mathrm{max}(\frac{1}{n} \sum_{i=1}^n S_{t_i} - K, 0)$. If we write $\frac{1}{n}(\sum_{i=1}^n S_{t_i} - K)$ in a different form, separating the still-unknown $S_{t_n}$ from the already-known $S_{t_1}, \ldots, S_{t_{n-1}}$, we can write the payoff as

$\frac{1}{n}\mathrm{max}\left( S_{t_n} - \left( nK - \sum_{i=1}^{n-1} S_{t_i} \right), 0 \right)$,

and since $nK - \sum_{i=1}^{n-1} S_{t_i}$ is known at time $t_{n-1}$, this is just a fractional multiple of the payoff function of a *vanilla* call with strike price $nK - \sum_{i=1}^{n-1} S_{t_i}$ and time to expiration $T - t_{n-1}$. Therefore, at this stage we could just use the Black-Scholes formula and skip simulating the final step. In our example above, this would allow us to replace $52 \times 52$ matrices with $51 \times 51$ matrices, which doesn't seem like much, but for an option with fewer look-at dates, being able to skip one step may be worth implementing in the code.
