# Mini Project 3

Throughout this mini-project, we assume that the standard Black-Scholes assumptions hold. We are, at time $t = 0$, considering a market which includes a *stock* whose value $t$ years from now is $S(t)$ dollars (the current value, $S(0)$, or *spot price*, we will often abbreviate to $S$) as well as a riskless money-market account that grows at the continuously compounded rate $r \geq 0$. We are interested in the price of a (European-style) *call option* $C$ struck at $K$ with expiration date $T$, which confers the right (but not the obligation) to purchase the stock at time $t = T$ for $K$ dollars. Because we are working with the Black-Scholes model, we assume that the stock price follows geometric Brownian motion with constant (log) drift $\mu$ and constant (log) volatility $\sigma$. The Black-Scholes formula, introduced and proved in the boot camp lectures, gives us the price of the call option $C$. This formula is

$C(S,T) = C(S, K, \sigma, r, T) = S\Phi(d_1) - Ke^{-rT}\Phi(d_2),$

where $\Phi(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^x e^{-z^2/2} \, dz$ is the cumulative distribution function (cdf) of the standard normal distribution, $d_1 = \frac{\ln(S/K) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma \sqrt{T}}$, and $d_2 = d_1 - \sigma \sqrt{T}$ (or, equivalently, what you get if you take the formula for $d_1$ and change the plus sign inside the parentheses to a minus sign). In the Black-Scholes model, this is the unique correct arbitrage-free price, in the sense that if you see $C$ trading in the market for any other price, you can lock in a risk-free profit by either buying $C$ (if it is underpriced) or selling (if overpriced), then carrying out a dynamic hedging procedure until expiration.

Even though the Black-Scholes price of $C$ depends on five quantities (six in the slightly more complicated version used when the stock $S$ pays dividends), we often abbreviate it to $C(S,T)$ because the stock price $S$ and time to expiration $T$ will change during the life of the option. In contrast, the strike price $K$ will not change, and the risk-free short rate $r$ and volatility $\sigma$ are assumed to be constant (they are parameters of the model).

The rate of change of a function of several variables as just one of these variables is changed (the others being assumed to be constant) is the **partial derivative** of that function with respect to that variable. So this mini-project is asking us to consider the partial derivatives of $C$ with respect to $T$ and $S$. In quantitative finance, the partial derivatives of the price of an option or other financial instrument are called **the Greeks**, and each is denoted by a different Greek letter:

- **Delta** is $\Delta = \partial C/\partial S$, the derivative with respect to the spot price;
- **Theta** is $\Theta = \partial C/\partial T$, the derivative with respect to time (one has to be careful here with signs, since as time increases, time to expiration decreases; this will be discussed further below);
- **Rho** is $\rho = \partial C/\partial r$, the derivative with respect to the short rate;
- **Vega** is $\nu = \partial C/\partial \sigma$, the derivative with respect to volatility.

"Vega" is not the name of an actual letter in the Greek alphabet. (The Greek letter $\nu$ (nu) looks something like *v*, the first letter of "volatility", in English.) To the best of my knowledge, there is not a Greek letter associated with $\partial C/\partial K$, the derivative with respect to strike. One can take repeated partial derivatives, or even multiple partial derivatives with respect to different variables, and many of these are studied and have been given Greek or pseudo-Greek names (like **Vanna** for $\partial^2 C/\partial \sigma \partial S$, the rate of change of Delta with respect to volatility). By far the most important and commonly studied higher partial derivative is **Gamma**, which is $\Gamma = \partial^2 C/\partial S^2$. This can be thought of as the rate of change of Delta with respect to the spot price.

In this language, the mini-project instructions are to study the Delta and Theta of call and put options in the Black-Scholes model. Depending on how one reads "how the rate of change... behaves as time progresses", it could also be argued that Gamma should be studied here as well. There are several ways to proceed:

1. **Pure reason.** We should be able to get a qualitative sense of whether an option's value goes up or down when the underlying stock's price increases by a dollar, or when another day passes, without using formulas at all but just thinking about what an option contract is and when it would be exercised. Obviously there are limits to how detailed the information we gain can be, but this is a good test of basic understanding.

    As an example of what we mean by (1) above, consider the difference between an **in the money (ITM)** and **out of the money (OTM)** call option. To be concrete, suppose the call option has a strike price of \\$100 and expires in one year. Now suppose further that the current value of the underlying stock (the spot price) is \\$75. As $S$ gets closer to $K$, the rate of change will increase, since the effect of one dollar's increase in $S$ on the probability that the option expires with positive value will be greater. As $S$ gets closer to $K$, the rate of change will increase, since the effect of one dollar's increase in $S$ on the probability that the option expires with positive value will be greater. The option's *intrinsic value* is zero, because if it were to expire today, you would not exercise it (it is OTM). Therefore, the only reason the option has a current value above zero is the chance that the spot price will rise above \\$100 in the next year. (In fact, the Black-Scholes price of this option assuming zero interest rates and 30% annualized volatility is just \\$2.32.) If we increase the spot price by a dollar (to \\$76), we have not increased the probability that the option will expire with positive value all that much, so therefore we do not expect the value of the option to change that much. Generalizing, if the option is "deep" out of the money ($S$ substantially less than $K$), then the rate of change of the call option price $C(S,T)$ with respect to $S$ is close to zero. As $S$ gets closer to $K$, the rate of change will increase, since the effect of one dollar's increase in $S$ on the probability that the option expires with positive value will be greater.

    Now suppose the strike and spot prices are reversed: the strike is \\$75, the spot price is \\$100, and all the other assumptions and parameters are the same. The option's *intrinsic value* is now \\$25: if it were to expire today, you would exercise it (it is ITM) and make a profit equal to the difference between the spot and strike (the optionality increases the value still more: the Black-Scholes price of such an option is \\$27.32). Now we view it as quite likely indeed that the option will expire with positive value. Each additional dollar increase in the spot price is (almost) another guaranteed dollar in our pocket at expiration. Generalizing, if the option is "deep" in the money ($S$ substantially greater than $K$), then the rate of change of the call option price $C(S,T)$ with respect to $S$ is close to $1$. So $\Delta$ is close to $0$ (respectively, $1$) and the graph is close to horizontal when $S << K$ (respectively, $S >> K$). It seems that we should have a logistic or sigmoid-shaped graph (where the steepest increase happens near $S = K$); the exact shape will become clearer once we draw some more precise pictures.

2. **Calculus.** We have an explicit Black-Scholes formula for the price of a call or put option. This formula includes the variables $S$ and $T$. So we can simply calculate the partial derivatives $\partial C/\partial S$, $\partial C/\partial T$, and $\partial^2 C/\partial S^2$. The answers are well-known and can be looked up, but we will give the details of the calculation for Delta as it is a worthwhile exercise with an aesthetically pleasing conclusion.

3. **Graphing.** Since we know the Black-Scholes formula, we can fix a starting set of values for the variables and parameters, then choose one of them to vary and plot the Black-Scholes price against the changing values of that parameter. We can also take the explicit formulas for the Greeks obtained as in (2) above and draw their graphs. This will show us whether our qualitative conclusions reached above in (1) for Delta make sense. Finally, we can build two-dimensional matrices and/or `seaborn` heatmaps that show the dependence of the call/put option price on spot price and time to expiration simultaneously.

4. **Finite differences.** The derivative is the limit of difference quotients, so we can approximate the values of the Greeks by computing the option price for two different (but close together) values of the parameter we are interested in varying, then taking the appropriate difference quotient. More precisely, in the case of a call option,

    $\Delta(S,T) = \frac{\partial C}{\partial S}(S,T) = \lim_{\varepsilon \rightarrow 0} \frac{C(S+\varepsilon,T) - C(S,T)}{\varepsilon}$,

    and if we plug in a specific small number for $\varepsilon$, we should get an approximation for $\Delta$ (Delta) at that specific choice of spot price $S$ and time $T$. As a concrete example, we could calculate the Black-Scholes prices $C(100,1)$ and $C(100.05,1)$ for two call options with the same strike, time to expiration ($T = 1$ year), etc., take the difference, and multiply by $20$ (that is, divide by $\varepsilon = 100.05-100 = 0.05$) to approximate $\Delta(100,1)$. We could then repeat the process, taking smaller and smaller $\varepsilon$s, and see what the difference quotients are converging to. To approximate $\Theta$ (Theta), we would instead leave the spot price $S$ fixed and consider two different values for $T$. Finally, a comment on $\Gamma$ (Gamma): as Gamma is a second derivative, simulating Gamma using finite differences would at first seem to require a difference quotient of difference quotients (or a previous analysis/approximation of the first derivative, Delta). However, there is a trick (exercise 5.11 from Walter Rudin's *Principles of Mathematical Analysis*): we have

    $\Gamma(S,T) = \frac{\partial^2 C}{\partial S^2}(S,T) = \lim_{\varepsilon \rightarrow 0} \frac{C(S+\varepsilon, T) + C(S-\varepsilon, T) - 2C(S,T)}{\varepsilon^2}$,

    so we can calculate the expression on the right directly for small values of $\varepsilon$. (To see why this works, note that the Taylor series expansions of $C(S+\varepsilon, T)$ and $C(S-\varepsilon, T)$ with respect to $S$ have the same constant and quadratic terms, but their linear terms have opposite signs and hence cancel when added.) In fact, this Taylor series argument gives us a better way to approximate first derivatives like Delta as well: we can use $\frac{C(S+\varepsilon,T) - C(S-\varepsilon,T)}{2\varepsilon}$ instead of the original difference quotient, which will converge to $\Delta(S,T)$ faster for small $\varepsilon$ because the terms in the numerator that are linear in $\varepsilon$ will cancel.

5. **Monte Carlo.** In (4) above, the idea is to calculate the price of the option when the spot price is a specific value $S$, calculate it again when the spot price is $S + \varepsilon$ for a small number $\varepsilon$, and use the difference between the two obtained values to approximate the instantaneous rate of change. "Calculate" here means "plug values into the Black-Scholes formula". An alternative approach would be to approximate the price of the option using a Monte Carlo simulation (twice, once for spot price $S$ and then again for spot price $S + \varepsilon$) then compute the difference quotient as above using the Monte Carlo option prices instead of the Black-Scholes values.

Below, we will give details (descriptions, calculations, code, and plots/pictures) for approaches (2) and (3) above. We also provide code for (5), simulating all the Greeks using Monte Carlo and finite differences (including the second-derivative "shortcut" described above for Gamma).

## Calculus

### Delta for a call option

Here we will calculate the Black-Scholes Delta of a call option, $\Delta = \frac{\partial C}{\partial S}$, using calculus and the Black-Scholes formula for $C(S,T)$. 

**Claim.** $\Delta(S,T) = \frac{\partial}{\partial S} C(S,T) = \Phi(d_1)$.

Recall that $C(S,T) = S\Phi(d_1) - Ke^{-rT}\Phi(d_2)$. If $d_1$ and $d_2$ were (mistakenly) treated as constants, then the only dependence on $S$ would be the factor in front of the first term, making this look like a one-line calculation (derivative of the first term is $1 \cdot \Phi(d_1)$ and derivative of the second term is zero). This is a mistake, and it is therefore somewhat surprising that the resulting answer is still correct. Instead, since $d_1$ and $d_2$ both depend on $S$, we will need the product rule, chain rule, and (since $\Phi$ is defined using an integral) the fundamental theorem of calculus to work out

$\Delta(S,T) = \frac{\partial}{\partial S}(S\Phi(d_1) - Ke^{-rT}\Phi(d_2)) = \frac{\partial}{\partial S} S\Phi(d_1) - \frac{\partial}{\partial S} Ke^{-rT}\Phi(d_2)$.

We calculate these two terms separately below.

#### First term

This term is $\frac{\partial}{\partial S} S\Phi(d_1)$, which by the product rule is equal to $\Phi(d_1) + S\frac{\partial}{\partial S} \Phi(d_1)$. By definition, $\Phi(d_1) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{d_1} e^{-z^2/2} \, dz$, so by the fundamental theorem of calculus together with the chain rule, $\frac{\partial}{\partial S} \Phi(d_1) = \frac{1}{\sqrt{2\pi}} e^{-d_1^2/2} \frac{\partial}{\partial S} d_1$ (the chain rule is needed here because the upper bound of the integral was not $S$, but $d_1$, which is a function of $S$).

Now recall that $d_1 = \frac{\ln(S/K) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma \sqrt{T}}$. If we write this as the sum of two fractions by splitting up the numerator, the second term does not depend on $S$ and therefore vanishes upon differentiation. So we have 

$\frac{\partial}{\partial S} d_1 = \frac{\partial}{\partial S} \frac{\ln(S/K)}{\sigma \sqrt{T}} = \frac{1}{\sigma \sqrt{T}} \cdot \frac{1}{S/K} \cdot \frac{1}{K} = \frac{1}{S\sigma \sqrt{T}}$, 

by the chain rule. (Since $d_2 = d_1 - \sigma \sqrt{T}$ and the term $\sigma \sqrt{T}$ does not depend on $S$, this same calculation also shows that $\frac{\partial}{\partial S} d_1 = \frac{1}{S\sigma \sqrt{T}}$, which we will use below.) Putting our calculations together, we have

$\frac{\partial}{\partial S} S\Phi(d_1) = \Phi(d_1) + S\frac{\partial}{\partial S} \Phi(d_1) = \Phi(d_1) + \frac{1}{\sigma \sqrt{2\pi T}} e^{-d_1^2/2}$,

since factors of $S$ appear in the numerator and denominator of the second term above and therefore cancel.

#### Second term

This term is $\frac{\partial}{\partial S} Ke^{-rT}\Phi(d_2) = Ke^{-rT} \frac{\partial}{\partial S} \Phi(d_2)$. Repeating the calculations above but for $d_2$ instead of $d_1$, we find 

$\frac{\partial}{\partial S} \Phi(d_2) = \frac{1}{\sqrt{2\pi}} e^{-d_2^2/2} \frac{\partial}{\partial S} d_2 = \frac{1}{S \sigma \sqrt{2\pi T}} e^{-d_2^2/2}$,

since (as observed above) we have $\frac{\partial}{\partial S} d_2 = \frac{\partial}{\partial S} d_1 = 1/(S \sigma \sqrt{T})$. To proceed further, we need to use the fact that $d_2 = d_1 - \sigma\sqrt{T}$, and therefore

$e^{-d_2^2/2} = e^{-(d_1 - \sigma \sqrt{T})^2/2} = e^{-d_1^2/2}e^{d_1\sigma \sqrt{T} - \frac{\sigma^2 T}{2}}$.

Because $d_1 = \frac{\ln(S/K) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma \sqrt{T}}$, we have $d_1 \sigma \sqrt{T} = \ln(S/K) + \left(r + \frac{\sigma^2}{2}\right)T$, and so $d_1\sigma \sqrt{T} - \frac{\sigma^2 T}{2} = \ln(S/K) + rT$. Exponentiating, $e^{d_1\sigma \sqrt{T} - \frac{\sigma^2 T}{2}} = e^{\ln(S/K) + rT} = \frac{S}{K}e^{rT}$, and putting our calculations together, we have 

$\frac{\partial}{\partial S} Ke^{-rT}\Phi(d_2) = Ke^{-rT} \frac{1}{S \sigma \sqrt{2\pi T}} e^{-d_1^2/2} \frac{S}{K}e^{rT}$.

#### Conclusion

All we need to do now is subtract our second term from our first:

$\Delta = \frac{\partial}{\partial S} C(S,T) = \left(\Phi(d_1) + \frac{1}{\sigma \sqrt{2\pi T}} e^{-d_1^2/2}\right) - \left(Ke^{-rT} \frac{1}{S \sigma \sqrt{2\pi T}} e^{-d_1^2/2} \frac{S}{K}e^{rT}\right) = \Phi(d_1),$

since $Ke^{-rT}$ multiplied by $\frac{S}{K}e^{rT}$ is just $S$.

### Other Greeks (Gamma, Theta, and put option versions)

Since we have calculated above that the Delta for a call option is $\Delta(S,T) = \Phi(d_1)$, we can differentiate once more (again using the fundamental theorem of calculus) to find that

$\Gamma(S,T) = \frac{\partial}{\partial S} \Phi(d_1) = \Phi'(d_1) \frac{\partial}{\partial S} d_1 = \frac{1}{\sqrt{2\pi}} e^{-d_1^2/2} \frac{1}{S\sigma \sqrt{T}} = \frac{e^{-d_1^2/2}}{S \sigma \sqrt{2 \pi T}}$ (or just $\frac{\Phi'(d_1)}{S \sigma \sqrt{T}}$).

Often $\varphi$ is used to denote $\Phi'$, the probability density function (pdf) of the standard normal distribution. 

Theta is an entirely separate calculation, requiring more careful algebra similar to that used above for Delta; we omit the calculations here, merely stating that

$\Theta(S,T) = \frac{\partial}{\partial T} C(S,T) = -\frac{S\Phi'(d_1)\sigma}{2\sqrt{T}} - rKe^{-rT}\Phi(d_2)$.

Now that we have formulas for the Delta, Gamma, and Theta of a call option, we can use put-call parity to quickly get formulas for the same Greeks in the case of a put option. Put-call parity states that $C - P = F = S - Ke^{-rT}$, where $F$ here stands for a *forward contract*. (The buyer of a forward contract struck at $K$ with time to expiration $T$ has the right **and** obligation to buy the stock at time $T$ for $K$ dollars. Since the seller of a forward contract can buy the stock today for $S$ dollars and will receive $K$ dollars at time $T$ (present value $Ke^{-rT}$ dollars today), they should receive $S - Ke^{-rT}$ additional dollars today for selling the contract in order to break even, which is where the formula for $F$ comes from. If someone is long a call option and short a put option with the same strike and expiration, then they will receive the stock and lose $K$ dollars at time $T$ for certain, meaning that their portfolio is equivalent to a forward. This is why $C - P = F$.) 

Since partial differentiation is linear, any Greek for $C$ minus that Greek for $P$ equals the same Greek for $F = S - Ke^{-rT}$. The Delta for $F$ is $\frac{\partial F}{\partial S} = 1$, and therefore the Gamma for $F$ is $\frac{\partial}{\partial S} 1 = 0$. This means that the Delta for a put option is the call Delta minus $1$ (that is, $\Phi(d_1) - 1$), and the Gamma for a put option is equal to the Gamma for a call option (since their difference is zero). Finally, the Theta for $F$ is $\frac{\partial F}{\partial T} = rKe^{-rT}$, so using the call Theta formula above, we find that the Theta for a put option must be $-\frac{S\Phi'(d_1)\sigma}{2\sqrt{T}} + rKe^{-rT}\Phi(-d_2)$ (since $\Phi(-d_2) = 1 - \Phi(d_2)$ by symmetry, you can check that the call and put option Thetas have difference equal to $rKe^{-rT}$).

## Graphing

We begin by writing Python functions that evaluate all the explicit formulas above (call and put Deltas, Gamma which is the same for calls and puts, and finally call and put Thetas). $\Phi$ and $\varphi$ above are calculated using `scipy.stats.norm.cdf` and `scipy.stats.norm.pdf` respectively.

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
import seaborn as sns
sns.set_style('darkgrid')

def bs_call(S_0, K, sigma, T = 1, r = 0):
    """
    Description:
    
    Computes the Black-Scholes value of a European call option.
    
    Parameters:
        S_0: Current asset price
        K: Strike price
        sigma: Yearly standard deviation of log-returns (volatility)
        T: Time to expiration (in years)
        r: Risk-free interest rate
    
    Returns:
        Call option price
    """
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    # d_2 = (np.log(S_0/K) + (r-.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d_2 = d_1 - sigma*np.sqrt(T)
    return S_0*norm.cdf(d_1)-K*np.exp(-r*T)*norm.cdf(d_2)
    
def bs_put(S_0, K, sigma, T = 1, r = 0):
    """
    Description:
    
    Computes the Black-Scholes value of a European put option.
    
    Parameters:
        same as in bs_call above
    
    Returns:
        Put option price
    """
    
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d_2 = d_1 - sigma*np.sqrt(T)
    return -S_0*norm.cdf(-d_1)+K*np.exp(-r*T)*norm.cdf(-d_2)

def bs_call_delta(S_0, K, sigma, T = 1, r = 0):
    """
    Computes the delta (sensitivity to changes in spot price) of a call option's price.
    Variables and parameters are the same as in bs_call above.
    (If the underlying stock price increases by one dollar, then the call option's price increases by approximately delta dollars.)
    """
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    return norm.cdf(d_1)

def bs_put_delta(S_0, K, sigma, T = 1, r = 0):
    """
    Computes the delta, this time for a put option.
    Variables and parameters are the same as in bs_call above.
    """
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    return norm.cdf(d_1) - 1

def bs_gamma(S_0, K, sigma, T = 1, r = 0):
    """
    Gamma for call and put options is the same.
    Variables and parameters are the same as in bs_call above.
    (If the underlying stock price increases by one dollar, then delta increases by approximately gamma dollars-per-dollar.)
    """
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d_2 = d_1 - sigma*np.sqrt(T)
    return (norm.pdf(d_1))/(S_0*sigma*np.sqrt(T))
    
def bs_call_theta(S_0, K, sigma, T = 1, r = 0):
    """
    Computes the theta (sensitivity to changes in time, a.k.a., time decay) of a call option's price.
    Variables and parameters are the same as in bs_call above.
    """
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d_2 = d_1 - sigma*np.sqrt(T)
    return (-(S_0*norm.pdf(d_1)*sigma)/(2*np.sqrt(T)) - r*K*np.exp(-r*T)*norm.cdf(d_2))/252 # per trading day
             
def bs_put_theta(S_0, K, sigma, T = 1, r = 0): 
    """
    Computes the theta, this time for a put option.
    Variables and parameters are the same as in bs_call above.
    """
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d_2 = d_1 - sigma*np.sqrt(T)
    return (-(S_0*norm.pdf(d_1)*sigma)/(2*np.sqrt(T)) + r*K*np.exp(-r*T)*norm.cdf(-d_2))/252 # per trading day
```

The unit of time $T$ in the Black-Scholes formula is the year. However, in practice, options traders quote Theta as the loss of value after a *day* has passed, so we have divided by $252$ (the usual number of trading days in a year) in the formulas above. It is also common to divide by $365$ instead, counting all days and not just trading days in the time decay calculation. We write "*loss* of value" because Theta is always negative for call options, which can be seen by looking at the explicit formula: as time passes, optionality is lost and hence the option becomes less valuable. (There are cases in which deep-in-the-money put options may have positive Theta.)

### Call option plots

Below, we draw graphs of the final option payoff (if the option were to expire today), option price, Delta, and Gamma plotted against the spot price for the specific case of a call option struck at \$100 and expiring in six months.

```python
spot_prices = np.linspace(50,150,100)
K = 100 # option struck at $100
sigma = .3 # 30% annualized volatility
T = 1/2 # option expires in six months

payoffs = [max(spot_prices[i] - K, 0) for i in range(len(spot_prices))]
call_prices = [bs_call(spot_prices[i], K, sigma, T) for i in range(len(spot_prices))]
call_deltas = [bs_call_delta(spot_prices[i], K, sigma, T) for i in range(len(spot_prices))]
call_gammas = [bs_gamma(spot_prices[i], K, sigma, T) for i in range(len(spot_prices))]

fig,axes = plt.subplots(2,2,figsize=(8,8))

axes[0,0].plot(spot_prices, payoffs)
axes[0,0].set_xlabel("Spot") 
axes[0,0].set_ylabel("Payoff") 
axes[0,0].set_title("Final call payoff as a function of spot")
axes[0,1].plot(spot_prices, call_prices)
axes[0,1].set_xlabel("Spot") 
axes[0,1].set_ylabel("Call Price") 
axes[0,1].set_title("Call price as a function of spot")
axes[1,0].plot(spot_prices, call_deltas)
axes[1,0].set_xlabel("Spot") 
axes[1,0].set_ylabel("Delta") 
axes[1,0].set_title("Delta as a function of spot (call)")
axes[1,1].plot(spot_prices, call_gammas)
axes[1,1].set_xlabel("Spot") 
axes[1,1].set_ylabel("Gamma") 
axes[1,1].set_title("Gamma as a function of spot")

plt.subplots_adjust(hspace = 0.25, wspace = 0.25)

plt.show()
```

### Observations (call option's sensitivity to spot price)

- Gamma is positive for all values of spot. This is clear from the formula $\Gamma(S,T) = \frac{\Phi'(d_1)}{S \sigma \sqrt{T}}$, since $\Phi'(d_1) = \int_{-\infty}^{d_1} e^{-z^2/2} \, dz$ is always positive, but it is still worth remarking on explicitly (it means that the price of a call option is a *convex* function of the spot price).

- Gamma is *not* symmetric about $S = K = 100$: the peak of Gamma's graph occurs a bit to the left of this value. Note also that Delta is not exactly $0.5$ at this value of $S$, but appears to be a bit larger. This is because if $S = K$, then $\ln(S/K) = 0$, so $d_1 = \frac{\left(r + \frac{\sigma^2}{2}\right)T}{\sigma \sqrt{T}} > 0$. Since the pdf $\varphi$ of the normal distribution is symmetric about zero, $d_1 > 0$ implies $\Phi(d_1) > 1/2$.

- In the introductory block above, we reasoned that Delta should be approximately flat at zero when the call option is deep out of the money, increase rapidly when spot is close to strike, and then level off, becoming approximately flat at one when the call option is deep in the money. The plot above shows that this is indeed the case.

- The second plot (the call option price) looks like a "rounded" version of the first plot (the final payoff as a function of spot). As the expiration date nears, the second graph will look more and more like the first one.

### Put option plots

Next, we draw the same plots (same parameter values) but for a put option instead.

### Observations (put option's sensitivity to spot price)

Recall that Gamma for a call and Gamma for a put are the same, so there is nothing new to say about the graph of Gamma. The shape of the graph for Delta is the same as it was for a call, but now the values of Delta are all negative (Delta is close to $-1$ when the put option is deep ITM and close to $0$ when the put option is deep OTM). This makes sense: a put option is a bet *against* the stock (or, more conservatively, it is insurance against a drop in its price), so when the stock price increases, the put option's value is never going to go up. If the put option is so deep in the money that it will surely be exercised, then every dollar increase in the price of the stock is now a dollar that will *not* be in our (the option holder's) pocket.

Again we see that the graph of the option's price looks something like a rounded version of the graph of its final payoff. An option with positive time to expiration is always worth more than its intrinsic value, so the option price graph is always strictly above the payoff graph (though the difference may be so small for deep OTM or ITM options that the naked eye cannot tell them apart on plots of this size).

### Theta

We now consider Theta, first for calls and then for puts. We will put the graph of the call option's price and the graph of its Theta side-by-side for an ITM, ATM (at the money), and OTM call where the horizontal axis gives a range of times to expiration (from one month to five years). The spot price in the plots below is \\$100, and the strike prices for the ITM, ATM, and OTM calls respectively are \\$85, \\$100, and \\$115.

### Observations (call option's sensitivity to time)

- Theta is always negative in all the plots above. This is called the "time decay" of the option: as time passes, you have less optionality (fewer rights) and so the contract you hold becomes less valuable (the bonus value of the contract above its intrinsic value shrinks to zero).
  
- As time to expiration approaches zero, the overall pattern is that Theta decreases (becomes more negative). For example, all three Theta plots above show this pattern between $T = 2$ and $T = 1$ (time decay is worse when there is only one year until expiration than when there are two years). However...
  
- ...when expiration is close, we see very different behavior in the ITM and OTM plots than in the ATM plot. Why is this?
  
- When the option is comfortably in or out of the money, market participants have a good sense of what is going to happen if expiration is soon. The ITM option is probably going to expire with positive value. The OTM option is probably going to expire worthless. The passage of one more day doesn't change our feelings about this too much.
  
- But when the option is at the money and expiration is soon, a swing one way in the stock price becomes less likely to be canceled out in the future.

### Appendix: Monte Carlo and finite differences (calculation of all Greeks) 

Below we implement approaches (4) and (5) from the introductory discussion to find the values of all Greeks (Delta, Gamma, Theta, Vega, and Rho) using Monte Carlo together with finite differences. We perform all simulations at once, returning an `nparray` with all five values, and then compare it against an array containing the Black-Scholes values. The Black-Scholes values for Delta, Gamma, and Theta were either computed or at least quoted above, but we have not mentioned the explicit formulas for Vega and Rho, which are $\nu(S,T) = S\varphi(d_1)\sqrt{T}$ and $\rho(S,T) = KTe^{-rT}\Phi(d_2)$.

*Note:* We did not scale by $1/252$ in the calculation of Theta below (we also did not scale Vega or Rho, which in practice are usually normalized to correspond to one percentage point increase and one "basis point"-- i.e., hundredth of a percentage point-- increase respectively).

```python
def bs_MC_call_greeks(S_0, K, sigma, t, r, n_sims = 500): # simulate Greeks using Monte Carlo
    
    """Description: 
    Monte Carlo Simulation of Black-Scholes call option Greeks (delta, gamma, theta, vega, and rho)
    
    Parameters:
    S_0 (float): spot price
    K (float): strike price
    sigma (float): volatility
    t (float): time to expiration
    r (float): risk-free interest rate
    n_sims (int): Number of simulations
    
    Return:
    nparray(float): Monte Carlo-calculated values of call option Greeks 
    """
    s_bump = .01*S_0 # "epsilon": in each case, we take the "bump" to be 1% of the original value

    v_bump = .01*sigma

    t_bump = .01*t

    r_bump = .01*r

    noise = np.random.normal(0,1,n_sims)

    log_returns = (r-.5*sigma**2)*t + sigma*np.sqrt(t)*noise

    paths = S_0*np.exp((r-.5*sigma**2)*t + sigma*np.sqrt(t)*noise)

    paths_delta_up = (S_0+s_bump)*np.exp((r-.5*sigma**2)*t + sigma*np.sqrt(t)*noise)

    paths_delta_down = (S_0-s_bump)*np.exp((r-.5*sigma**2)*t + sigma*np.sqrt(t)*noise)

    paths_theta_up = S_0*np.exp((r-.5*sigma**2)*(t+t_bump) + sigma*np.sqrt(t+t_bump)*noise)

    paths_theta_down = S_0*np.exp((r-.5*sigma**2)*(t-t_bump) + sigma*np.sqrt(t-t_bump)*noise)

    paths_vega_up = S_0*np.exp((r-.5*(sigma+v_bump)**2)*t + (sigma+v_bump)*np.sqrt(t)*noise)

    paths_vega_down = S_0*np.exp((r-.5*(sigma-v_bump)**2)*t + (sigma-v_bump)*np.sqrt(t)*noise)

    paths_rho_up = S_0*np.exp((r+r_bump-.5*sigma**2)*t + sigma*np.sqrt(t)*noise)

    paths_rho_down = S_0*np.exp((r-r_bump-.5*sigma**2)*t + sigma*np.sqrt(t)*noise)

    delta_sims = ((np.maximum(paths_delta_up - K, 0) - np.maximum(paths_delta_down - K, 0))*np.exp(-r*t))/(2*s_bump)

    gamma_sims = ((np.maximum(paths_delta_up - K, 0) + np.maximum(paths_delta_down - K, 0) - 2*np.maximum(paths - K, 0))*np.exp(-r*t))/(s_bump**2)

    vega_sims = ((np.maximum(paths_vega_up - K, 0) - np.maximum(paths_vega_down - K, 0))*np.exp(-r*t))/(2*v_bump)

    # for theta and rho, the discount factor changes depending on whether we bumped up or down

    theta_sims = ((np.maximum(paths_theta_up - K, 0)*np.exp(-r*(t + t_bump)) - np.maximum(paths_theta_down - K, 0)*np.exp(-r*(t - t_bump))))/(2*t_bump)

    rho_sims = ((np.maximum(paths_rho_up - K, 0)*np.exp(-(r + r_bump)*t) - np.maximum(paths_rho_down - K, 0)*np.exp(-(r - r_bump)*t)))/(2*r_bump)

    delta_estimate = np.mean(delta_sims)

    gamma_estimate = np.mean(gamma_sims)

    vega_estimate = np.mean(vega_sims)

    theta_estimate = -np.mean(theta_sims) # sign conventions! as written above, theta is positive, while the B-S formula gives a negative value
                                          # often theta is quoted as a positive number-- we just have to be careful about conventions

    rho_estimate = np.mean(rho_sims)

    return np.array([delta_estimate, gamma_estimate, vega_estimate, theta_estimate, rho_estimate])


def bs_call_greeks(S_0, K, sigma, t, r): # explicit formulas
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d_2 = d_1 - sigma*np.sqrt(t)
      
    bs_call_delta = norm.cdf(d_1)
    bs_gamma = norm.pdf(d_1)/(S_0*sigma*np.sqrt(t))
    bs_vega = S_0*norm.pdf(d_1)*np.sqrt(t)
    bs_call_theta = (-S_0*norm.pdf(d_1)*sigma)/(2*np.sqrt(t)) - r*K*np.exp(-r*t)*norm.cdf(d_2)
    bs_call_rho = K*t*np.exp(-r*t)*norm.cdf(d_2)
    return np.array([bs_call_delta, bs_gamma, bs_vega, bs_call_theta, bs_call_rho])

print(bs_MC_call_greeks(100,105,0.25,1,0.03,100).round(3))
print(bs_MC_call_greeks(100,105,0.25,1,0.03,10000).round(3))
print(bs_MC_call_greeks(100,105,0.25,1,0.03,1000000).round(3))
print(bs_call_greeks(100,105,0.25,1,0.03).round(3))
```
