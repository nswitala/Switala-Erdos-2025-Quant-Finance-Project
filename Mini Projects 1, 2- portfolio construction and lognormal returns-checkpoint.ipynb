# Mini Project 1
*"Use current stock data to create two potentially profitable investment portfolios. One that is higher risk and one that is lower risk. (You are to interpret and explain your interpretation of a high risk profile and low risk profile of a portfolio. You should provide some measurable quantitative data in your explanation.)"*

The **Global Industry Classification Standard (GICS)**, developed in 1999 by MSCI and Standard & Poor's, classifies all S&P 500 stocks as belonging to one of eleven *sectors:* information technology, health care, financials, consumer discretionary, consumer staples, energy, real estate, materials, utilities, communication services, and industrials.

**Portfolio 1**, which we will call `riskier_portfolio` in the Python code, will consist of the top ten stocks in the *information technology sector only*, ranked by market capitalization (or *cap*, i.e., the number of outstanding shares multiplied by the current price of a share). All market cap information used to build the portfolios for this project was accessed during Memorial Day weekend 2025. The portfolio will give equal weight to all ten stocks; if the notional amount to invest in the portfolio is 10,000 USD, then we simply invest 1,000 USD in each stock (this likely requires owning fractional amounts of shares of some or all stocks in the portfolio to make the total value exactly 1,000, but these days even retail brokerages like Fidelity and Charles Schwab offer fractional shares). As we will be interested in rate of return, not absolute dollar amount, there will in practice not be a need to work explicitly with the 10,000 USD notional.

Portfolio 1 represents a targeted bet on one sector which has a historical reputation for being highly volatile. This impression will be confirmed below when we compare the recent volatilities of the largest stocks in the sector with those in other sectors.

The ten stocks making up Portfolio 1 are the following: **MSFT** (Microsoft), **NVDA** (Nvidia), **AAPL** (Apple), **AVGO** (Broadcom), **ORCL** (Oracle), **PLTR** (Palantir), **CRM** (Salesforce), **CSCO** (Cisco), **IBM** (IBM), and **INTU** (Intuit).

**Portfolio 2**, which we will call `safer_portfolio` in the Python code, will consist of one stock in each of the eleven sectors. (As with portfolio 1, all stocks in portfolio 2 will have equal weight.) The stock chosen to represent the sector will be the one which, among the five stocks in the sector with the largest market caps, had the *lowest* volatility (annualized standard deviation of logarithmic daily returns) for the first quarter of 2025. (Here we mean first *calendar* quarter, i.e., 1/1/2025-3/31/2025, not first *fiscal* quarter.) These volatilities are calculated below using data accessed from `yfinance`, where we find that the eleven stocks making up Portfolio 2 are the following: **MSFT** (Microsoft), **JNJ** (Johnson & Johnson), **V** (Visa), **TJX** (TJX Companies), **KO** (Coca-Cola), **CVX** (Chevron), **WELL** (Welltower), **LIN** (Linde), **DUK** (Duke Energy), **T** (AT&T), and **HON** (Honeywell).

By choosing large stocks with lower recent volatility from across the full range of market sectors, Portfolio 2 aims to minimize risk. One could argue that a portfolio with only 11 stocks in it is not nearly diversified enough for an investor to breathe easily, but the goal is to compare portfolios 1 and 2 directly, to see the difference between betting on one sector and spreading capital across sectors even when the number of stocks involved remains small. The goal in constructing Portfolio 2 is not to approximate the S&P 500.

This brings us to...

**Portfolio 3**, which we will call `baseline_portfolio` in the Python code. This portfolio consists of only two assets: **VOO** (Vanguard Group's exchange-traded fund, or ETF, that tracks the performance of the S&P 500) and **BRK-B** (Berkshire Hathaway's class B shares). The weighting of these two assets in Portfolio 3 is 75% VOO and 25% BRK-B. Berkshire Hathaway has significant stakes in many U.S. companies, such as Coca-Cola, and owns others outright, such as the insurer GEICO. A portfolio consisting of an S&P 500 index fund together with Berkshire can be thought of as a bet on the overall market with slightly heavier weight placed on companies in which Berkshire invests, which famously has excluded nearly all "tech stocks". The reason for including Portfolio 3 as a point of comparison rather than simply the S&P 500 is personal and perhaps silly: it was the author's own first portfolio. 

Looking below at the top five stocks in each sector and their volatilities, we see that annualized volatilities between the high teens and low thirties are typical. TSLA (in the consumer discretionary sector) and CEG (in utilities) are outliers, with very high volatility. However, the information technology sector stands out, with three out of the top five largest-cap stocks above 50\% annualized volatility. This is why this sector was chosen for the "risky" Portfolio 1.

**Note:** the first three months of 2025 were particularly volatile. The wisdom of using data from these months to construct Portfolio 2 may therefore be called into question. However, as the goal was to choose a large, comparatively "safe" stock from each sector, it could also be argued that looking for companies that remain somewhat steady even in turbulent times is a good idea.

Below, we investigate what would have happened to $10,000 invested in each of Portfolios 1, 2, and 3 on 1/1/2025 as of 3/31/2025. We also calculate the annualized volatility (standard deviation of logarithmic daily returns) for each portfolio.

```python
riskier_portfolio_tickers = ['MSFT', 'NVDA', 'AAPL', 'AVGO', 'ORCL', 'PLTR', 'CRM', 'CSCO', 'IBM', 'INTU']
start_date = dt.date(2025,1,1)
end_date = dt.date(2025,3,31)

# load data for these ten stocks for the first three months of 2025
riskier_portfolio_table = yf.download(riskier_portfolio_tickers, start = start_date, end = end_date)

# create a DataFrame of daily returns of all ten stocks
riskier_portfolio_returns = riskier_portfolio_table['Close']/riskier_portfolio_table['Close'].shift(1)
riskier_portfolio_returns = riskier_portfolio_returns.dropna()
riskier_aggregate_returns = np.mean(riskier_portfolio_returns, axis = 1) 

# the previous line deserves some explanation
# if prices (p_1, ..., p_10) become prices (q_1, ..., q_10) after one trading day,
# then the respective return factors are q_1/p_1, ..., q_10/p_10
# an equal-weight portfolio (with N/10 dollars invested in each stock, or N dollars total) would grow to N(q_1/p_1 + ... q_10/p_10),
# which is the same as N/10((q_1/p_1 + ... + q_10/p_10)/10), where the number in parentheses is the arithmetic mean of the return factors

riskier_aggregate_log_returns = np.log(riskier_aggregate_returns) # log returns to be used for calculating volatility
riskier_overall_return = np.prod(riskier_aggregate_returns) # multiply the starting amount invested by this product to get the amount you end with

# annualize the volatility (standard deviation of log returns) by multiplying by sqrt(250)
# there are 250 trading days in 2025 (252 is more typical)
riskier_annualized_volatility = np.std(riskier_aggregate_log_returns)*np.sqrt(250)
print(f"If USD 10000 were invested in Portfolio 1 on January 1, 2025, the portfolio's value at the close of trading on March 31, 2025, would be "
f"USD {10000*riskier_overall_return}. The annualized volatility of this portfolio is {riskier_annualized_volatility}%.")
```

Portfolio 2, consisting of only eleven equally-weighted stocks, was *less* volatile than Portfolio 3 (S&P 500 index fund + Berkshire Hathaway). However, Portfolio 3 had a slightly better rate of return. 

```python
# here we simulate what happens when $10,000 is invested in each of the portfolios and plot the results
# we multiply yesterday's value by today's return factor to obtain today's value

riskier_value = [10000]
safer_value = [10000]
baseline_value = [10000]
for i in range(len(riskier_aggregate_returns)):
    riskier_value.append(riskier_value[i]*riskier_aggregate_returns.values[i])
    safer_value.append(safer_value[i]*safer_aggregate_returns.values[i])
    baseline_value.append(baseline_value[i]*baseline_aggregate_returns[i]) 
# because of the weighted average step above, riskier/safer_aggregate_returns are different types of object (pandas Series vs. numpy ndarray)
# than baseline_aggregate_returns

plt.figure(figsize = (12,8))

plt.plot(riskier_portfolio_returns.index, riskier_value[1:], color = 'r', label = 'Portfolio 1')
plt.plot(safer_portfolio_returns.index, safer_value[1:], color = 'b', label = 'Portfolio 2')
plt.plot(baseline_portfolio_returns.index, baseline_value[1:], color = 'y', label = 'Portfolio 3')
plt.title('2025 Q1 Performance of Portfolios 1, 2, and 3', size = 16 )
plt.legend()
plt.show()
```

*"In the theory of mathematical finance, it is common to assume the log returns of a stock/index are normally distributed. Investigate if the log returns of stocks or indexes of your choosing are normally distributed."*

As mentioned above, the first three months of 2025 were particularly volatile. Therefore, to check whether our three portfolios have had lognormal distributions, we will look at the quarters of 2024 instead. For each of the four quarters of 2024 (January-March, April-June, July-September, and October-December; note that I do not mean *fiscal* quarters here), we will apply the Shapiro-Wilk test.

The null hypothesis of the Shapiro-Wilk test is that the data (here, the daily log returns) *are* normally distributed. A low $p$-value therefore provides evidence that the data are *not* normally distributed. We will use $p < 0.05$ as our threshold. Therefore, if the Shapiro-Wilk $p$ is greater than or equal to $0.05$, we have some confidence that the log returns *are* normally distributed.

The Shapiro-Wilk test is not the only test for normality. It is considered one of the more powerful tests, but its weakness is that, for large sample sizes, it can be oversensitive to small deviations from normality. Therefore, we will look below at one quarter at a time, rather than a full year's data. 

At the end, we will also look at the full year's data, and repeat the calculations from Mini Project 1 (we will investigate what would have happened to $10,000 invested in each of Portfolios 1, 2, and 3 on 1/1/2024 as of the end of the year, and also calculate the volatility of the portfolios' log returns during the whole year).

```python
riskier_portfolio_tickers = ['MSFT', 'NVDA', 'AAPL', 'AVGO', 'ORCL', 'PLTR', 'CRM', 'CSCO', 'IBM', 'INTU']
safer_portfolio_tickers = ['MSFT', 'JNJ', 'V', 'TJX', 'KO', 'CVX', 'WELL', 'LIN', 'DUK', 'T', 'HON']
baseline_portfolio_tickers = ['VOO', 'BRK-B']
start_date = dt.date(2024,1,1)
end_date = dt.date(2024,3,31)

riskier_portfolio_table = yf.download(riskier_portfolio_tickers, start = start_date, end = end_date)
riskier_portfolio_returns = riskier_portfolio_table['Close']/riskier_portfolio_table['Close'].shift(1)
riskier_portfolio_returns = riskier_portfolio_returns.dropna()
riskier_aggregate_returns = np.mean(riskier_portfolio_returns, axis = 1) 
riskier_aggregate_log_returns = np.log(riskier_aggregate_returns)

safer_portfolio_table = yf.download(safer_portfolio_tickers, start = start_date, end = end_date)
safer_portfolio_returns = safer_portfolio_table['Close']/safer_portfolio_table['Close'].shift(1)
safer_portfolio_returns = safer_portfolio_returns.dropna()
safer_aggregate_returns = np.mean(safer_portfolio_returns, axis = 1) 
safer_aggregate_log_returns = np.log(safer_aggregate_returns)

baseline_portfolio_table = yf.download(baseline_portfolio_tickers, start = start_date, end = end_date)
baseline_portfolio_returns = baseline_portfolio_table['Close']/baseline_portfolio_table['Close'].shift(1)
baseline_portfolio_returns = baseline_portfolio_returns.dropna()
baseline_aggregate_returns = np.average(baseline_portfolio_returns, axis = 1, weights = [0.75, 0.25]) 
baseline_aggregate_log_returns = np.log(baseline_aggregate_returns)

# Collect p-values of normality tests
p_1 = stats.shapiro(riskier_aggregate_log_returns)[1]     
p_2 = stats.shapiro(safer_aggregate_log_returns)[1]
p_3 = stats.shapiro(baseline_aggregate_log_returns)[1]

# Print evidence/non-evidence of normality
print(f"Portfolio 1 log return distribution: p-value = {p_1:.3f}")
if p_1 < 0.05:
    print("→ Statistically significant evidence that the data is NOT normally distributed.")
else:
    print("→ No statistically significant evidence against normality.")
    
print('--'*40) 
print('--'*40) 

print(f"Portfolio 2 log return distribution: p-value = {p_2:.3f}")
if p_2 < 0.05:
    print("→ Statistically significant evidence that the data is NOT normally distributed.")
else:
    print("→ No statistically significant evidence against normality.")

print('--'*40) 
print('--'*40) 

print(f"Portfolio 3 log return distribution: p-value = {p_3:.3f}")
if p_3 < 0.05:
    print("→ Statistically significant evidence that the data is NOT normally distributed.")
else:
    print("→ No statistically significant evidence against normality.")
```

```python
riskier_portfolio_tickers = ['MSFT', 'NVDA', 'AAPL', 'AVGO', 'ORCL', 'PLTR', 'CRM', 'CSCO', 'IBM', 'INTU']
safer_portfolio_tickers = ['MSFT', 'JNJ', 'V', 'TJX', 'KO', 'CVX', 'WELL', 'LIN', 'DUK', 'T', 'HON']
baseline_portfolio_tickers = ['VOO', 'BRK-B']
start_date = dt.date(2024,1,1)
end_date = dt.date(2024,12,31)

riskier_portfolio_table = yf.download(riskier_portfolio_tickers, start = start_date, end = end_date)
riskier_portfolio_returns = riskier_portfolio_table['Close']/riskier_portfolio_table['Close'].shift(1)
riskier_portfolio_returns = riskier_portfolio_returns.dropna()
riskier_aggregate_returns = np.mean(riskier_portfolio_returns, axis = 1) 
riskier_aggregate_log_returns = np.log(riskier_aggregate_returns)
riskier_overall_return = np.prod(riskier_aggregate_returns) 
riskier_annualized_volatility = np.std(riskier_aggregate_log_returns)*np.sqrt(251) #2024 had 251 trading days
p_1 = stats.shapiro(riskier_aggregate_log_returns)[1]  

print(f"If USD 10000 were invested in Portfolio 1 on January 1, 2024, the portfolio's value at the close of trading on December 31, 2024, would be "
f"USD {10000*riskier_overall_return}. The annualized volatility of this portfolio is {riskier_annualized_volatility}%.")

print(f"Portfolio 1 log return distribution: p-value = {p_1:.3f}")
if p_1 < 0.05:
    print("→ Statistically significant evidence that the 2024 Portfolio 1 data is NOT normally distributed.")
else:
    print("→ No statistically significant evidence against normality for Portfolio 1 in 2024.")

safer_portfolio_table = yf.download(safer_portfolio_tickers, start = start_date, end = end_date)
safer_portfolio_returns = safer_portfolio_table['Close']/safer_portfolio_table['Close'].shift(1)
safer_portfolio_returns = safer_portfolio_returns.dropna()
safer_aggregate_returns = np.mean(safer_portfolio_returns, axis = 1) 
safer_aggregate_log_returns = np.log(safer_aggregate_returns)
safer_overall_return = np.prod(safer_aggregate_returns) 
safer_annualized_volatility = np.std(safer_aggregate_log_returns)*np.sqrt(251)
p_2 = stats.shapiro(safer_aggregate_log_returns)[1]  

print(f"If USD 10000 were invested in Portfolio 2 on January 1, 2024, the portfolio's value at the close of trading on December 31, 2024, would be "
f"USD {10000*safer_overall_return}. The annualized volatility of this portfolio is {safer_annualized_volatility}%.")

print(f"Portfolio 2 log return distribution: p-value = {p_2:.3f}")
if p_2 < 0.05:
    print("→ Statistically significant evidence that the 2024 Portfolio 2 data is NOT normally distributed.")
else:
    print("→ No statistically significant evidence against normality for Portfolio 2 in 2024.")

baseline_portfolio_table = yf.download(baseline_portfolio_tickers, start = start_date, end = end_date)
baseline_portfolio_returns = baseline_portfolio_table['Close']/baseline_portfolio_table['Close'].shift(1)
baseline_portfolio_returns = baseline_portfolio_returns.dropna()
baseline_aggregate_returns = np.average(baseline_portfolio_returns, axis = 1, weights = [0.75, 0.25]) 
baseline_aggregate_log_returns = np.log(baseline_aggregate_returns)
baseline_overall_return = np.prod(baseline_aggregate_returns) 
baseline_annualized_volatility = np.std(baseline_aggregate_log_returns)*np.sqrt(251)
p_3 = stats.shapiro(baseline_aggregate_log_returns)[1]  

print(f"If USD 10000 were invested in Portfolio 3 on January 1, 2024, the portfolio's value at the close of trading on December 31, 2024, would be "
f"USD {10000*baseline_overall_return}. The annualized volatility of this portfolio is {baseline_annualized_volatility}%.")

print(f"Portfolio 3 log return distribution: p-value = {p_3:.3f}")
if p_3 < 0.05:
    print("→ Statistically significant evidence that the 2024 Portfolio 3 data is NOT normally distributed.")
else:
    print("→ No statistically significant evidence against normality for Portfolio 3 in 2024.")
```

Observe the large rate of growth of Portfolio 1 relative to the others during 2024, the "flip side" of Portfolio 1's losses during the first quarter of 2025. Its volatility during 2024 is also significantly greater than that of the other two portfolios.

Finally, we plot the performance of the three portfolios (with a $10,000 notional starting value) over the entire year 2024.

```python
riskier_value = [10000]
safer_value = [10000]
baseline_value = [10000]
for i in range(len(riskier_aggregate_returns)):
    riskier_value.append(riskier_value[i]*riskier_aggregate_returns.values[i])
    safer_value.append(safer_value[i]*safer_aggregate_returns.values[i])
    baseline_value.append(baseline_value[i]*baseline_aggregate_returns[i])

plt.figure(figsize = (12,8))

plt.plot(riskier_portfolio_returns.index, riskier_value[1:], color = 'r', label = 'Portfolio 1')
plt.plot(safer_portfolio_returns.index, safer_value[1:], color = 'b', label = 'Portfolio 2')
plt.plot(baseline_portfolio_returns.index, baseline_value[1:], color = 'y', label = 'Portfolio 3')
plt.title('2024 Performance of Portfolios 1, 2, and 3', size = 16 )
plt.legend()
plt.show()
```
