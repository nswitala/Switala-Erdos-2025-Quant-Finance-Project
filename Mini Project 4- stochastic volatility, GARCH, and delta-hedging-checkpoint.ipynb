# Mini Project 4

*"For this final mini project explore the impact of a non-constant $\sigma$ on the distribution of profits of hedging."*

Mini Projects 3 and "3.5" both relied heavily on the Black-Scholes model, in which the volatility $\sigma$ of the underlying asset (stock) is treated as a constant throughout the life of the option. It only takes minimal exposure to real-world stock and option data to be convinced that this assumption is not borne out in the markets. Therefore, a more realistic model will incorporate a dynamic $\sigma$ term that evolves with time and has a random component (**stochastic volatility**). In the boot camp lectures, we have encountered the **Heston model**, in which the stock's price $S_t$ and the price's instantaneous variance $v_t$ follow a *pair* of stochastic differential equations (SDE), the SDE for $v_t$ incorporates a mean reversion term (as in an Ornstein-Uhlenbeck process), and the Wiener processes providing the stochastic components of $S_t$ and $v_t$'s evolutions are assumed to have constant correlation.

In this mini-project, we will consider an alternative model for volatility, a **GARCH(1,1)** model. "GARCH" stands for *generalized autoregressive conditional heteroskedasticity*, meaning that the variance (volatility-squared) at a given time will depend on the previous variance as well as *its* variance, the (square of the) "vol-vol" or "volatility of volatility". (The exotic word at the end of the acronym-- also spelled *heteroscedasticity* in some sources-- simply means that the variances are not the same at all times, and ultimately derives from the Greek for "scattering".)

*Note:* I will assume throughout what follows that all "white noise" processes are Gaussian (that is, normally distributed). The definition of GARCH processes below can be adapted to use other distributions; the term GARCH does not imply in and of itself that normal distributions are being used.

A sequence $(\varepsilon_t)$ of residuals (error/noise terms) follows a **GARCH(p,q)** process if:
- there exist constants $\omega$, $\alpha_1, \ldots, \alpha_q$, and $\beta_1, \ldots, \beta_p$ such that
- $\sigma_t^2 = \omega + \sum_{i=1}^q \alpha_i \varepsilon_{t-i}^2 + \sum_{j=1}^p \beta_j \sigma_{t-j}^2$ for all $t \geq 1$ (terms with negative subscripts are declared zero), and
- $\varepsilon_t$, conditioned on all previously known terms, has an $N(0, \sigma_t^2)$ distribution.

When fitting a GARCH process to data, there are statistical tests (such as those based on the Akaike or Bayesian information criteria) one can perform to determine which lag lengths $p$ and $q$ are most suitable. However, we will focus here on the case $p = q = 1$ only (a GARCH(1,1) model), as this is the most commonly used in financial applications. (According to the V-Lab project at NYU Stern, *"V-Lab uses $p = 1$ and $q = 1$ though, because this is usually the option that best fits financial time series."*) We will also change the notation above somewhat, because the $\varepsilon$ above corresponds to the $\sigma$ from the Black-Scholes model that we wish to generalize, while the $\sigma$ above will be the "volatility of volatility" in our model. (We will still use geometric Brownian motion for our stock price, with drift equal to the risk-free interest rate, but we will now use a sequence of values for the volatility that follow a GARCH(1,1) process.) Since $p = q = 1$, we have only one $\alpha$ parameter and one $\beta$, and in our notation, $\zeta_t = \omega + \alpha \sigma_{t-1}^2 + \beta \zeta_{t-1}^2$ for all $t \geq 1$ (and $\sigma_t$ has an $N(0,\zeta_t^2)$ distribution).

There is no *closed* form for the price of a call or put option using the Heston model, but there is a "semi-closed form": the characteristic function of the stock price has an explicit form, and therefore the price itself can be found using numerical integration. Option prices under the Heston model can therefore be found efficiently, and the parameters of this model can be calibrated using option prices observed in the market. The strength of the Heston model is in its ability to price options and simulate a strategy in which options are possibly sold before expiration. GARCH models for volatility do not have this feature, so simulations are slower. The parameters in a GARCH model can be estimated using maximum likelihood (as we will do below). GARCH models are useful when modeling risk and volatility dynamics following a shock. All this is to say that we are not "taking sides" between the Heston model and GARCH models here; both have their uses, and we have chosen to look at GARCH here to explore something not covered in the boot camp lectures.

Below, we are going to simulate a portfolio consisting of a call option on the S&P 500 index together with a position in the index itself (for simplicity, consider holding an index-tracking ETF like SPY or VOO) used to periodically delta-hedge the option. In the simulation, we will use a GARCH(1,1) process to model the volatility of the index. We begin by downloading the previous year's S&P values from `yfinance` and calculating the daily log returns.

```python
sp500_index = yf.download('^GSPC', period = '1y', interval = '1d')
sp500_returns = sp500_index['Close']/sp500_index['Close'].shift(1)
sp500_returns = sp500_returns.dropna()
sp500_returns = np.log(sp500_returns)
```

Next, we use the previous year's data to calibrate the parameters of our GARCH(1,1) model. It is important to note that a GARCH process uses discrete time steps. We are calibrating the parameters based on daily historical return data, so the time step in our GARCH process will be equal to one day ($\sigma_t$ vs. $\sigma_{t-1}$ represents a change over a period of one day). When we use this information to simulate stock and option prices later, our time steps will be days and we will need to multiply/divide by $252$ as appropriate when considering time to option expiration measured in years.

From `yfinance` we have obtained a sequence $(\sigma_t)_{t=1}^N$ of daily log returns, where $N$ is the number of trading days in the previous year. We want to know which choice of GARCH(1,1) parameters $\omega$, $\alpha$, and $\beta$ maximize the *likelihood* that a GARCH process with those three parameters would produce the data we observe. Recall that the probability density function (pdf) of a normally distributed random variable with mean $0$ and variance $\zeta^2$ is $f(x) = \frac{1}{\zeta \sqrt{2 \pi}}e^{-x^2/2\zeta^2}$. The **likelihood function** of the parameters $\omega, \alpha, \beta$ with respect to the observed data $\vec{\sigma} = (\sigma_t)_{t=1}^N$ is

$L(\omega, \alpha, \beta | \vec{\sigma}) = \prod_{t=1}^N \frac{1}{\zeta_t \sqrt{2 \pi}}e^{-\sigma_t^2/2\zeta_t^2}$

where the $\zeta_t$ are defined using the GARCH(1,1) formulas $\zeta_t = \omega + \alpha \sigma_{t-1}^2 + \beta \zeta_{t-1}^2$ for all $t \geq 1$. (The likelihood function is just a product of pdfs; the conceptual difference here is that we are treating $L$ as a *function* of the unknown parameters. When passing from a probability density/mass function to a likelihood function, the observations and the parameters switch places.)

We want to find $\omega, \alpha, \beta$ such that $L(\omega, \alpha, \beta | \vec{\sigma})$ is maximal. Because the natural logarithm is an increasing function, it is equivalent to maximize the **log likelihood** 

$\ln L(\omega, \alpha, \beta | \vec{\sigma}) = \sum_{t=1}^N \ln \left(\frac{1}{\zeta_t \sqrt{2 \pi}} \right) - \frac{\sigma_t^2}{2\zeta_t^2}$;

since $\ln(1/\zeta_t) = -\ln(\zeta_t)$ and the natural logarithms of $1/\sqrt{2 \pi}$ and $1/2$ are constant terms that do not affect the problem, it is equivalent to maximize the objective function

$\sum_{t=1}^N \left( -\ln(\zeta_t) - \frac{\sigma_t^2}{\zeta_t^2} \right)$

which is the same thing as minimizing its negative, $\sum_{t=1}^N \left( \ln(\zeta_t) + \frac{\sigma_t^2}{\zeta_t^2} \right)$. This last remark is needed because we are going to use `scipy`'s `optimize.minimize` method to find this minimum numerically (and `scipy` does not have an analogous `maximize` method). This process is carried out in the code below. `scipy.optimize.minimize` requires an "initial guess" for the parameters to be optimized. Our initial guess, $(\omega, \alpha, \beta) = (0.000001, 0.1, 1)$, was obtained by rounding off the numbers found in a similar example in section 23.5 of John Hull's *Options, Futures, and Other Derivatives* (10th edition).

```python
N = len(sp500_returns)

def garch_log_likelihood(variables):
    """
    Using the sp500_returns data from above (a DataFrame of daily log returns),
    find the GARCH(1,1) parameters omega, alpha, beta that maximize the log likelihood of these returns.
    """
    #variables = [omega, alpha, beta]
    
    omega, alpha, beta = variables
    objective = 0 
    previous_var = sp500_returns.iloc[0,0]**2
    current_var = 0
    for i in range(1, N):
        current_var = omega + alpha*sp500_returns.iloc[i,0]**2 + beta*previous_var # no need to store all sigmas and zetas at once, since GARCH(1,1) 
                                                                                   # has a lag length of one
        objective += np.log(current_var) + (sp500_returns.iloc[i,0]**2)/current_var # add another term to the log-likelihood function (a running total)
        previous_var = current_var
    return objective

bounds = ((0, None), (0, None), (0, None)) # omega, alpha, and beta must be nonnegative

initial_guess = [0.000001, 0.1, 1.0] # rounded-off example from Hull's book

minimize(garch_log_likelihood, initial_guess, bounds = bounds)
```

Our conclusion is that $(\omega, \alpha, \beta) = (0.00000331, 0.08228, 0.8228)$ maximize the log-likelihood and therefore the likelihood. We will use these parameters in our GARCH(1,1) process below.

In the boot camp lectures, we learned to use *delta-hedging control variates* to reduce the variance in a Monte Carlo simulation of the price of a call option. The code took an additional argument, `n_hedges`, representing the number of periods during the life of the option when delta-hedging is to occur (a quantity of the underlying stock is bought or sold short in order to make the total number of shares held equal to the value of $\Delta = \frac{\partial C}{\partial S}$ at the given time, where $C$ and $S$ represent the price of the call option and the price of the stock respectively). 

Below, the code from the Jupyter notebooks accompanying these lectures is adapted to use our GARCH model for volatility instead of the Black-Scholes constant. 
- As mentioned above, the unit of time used to define the GARCH process is the day. Our simulation takes an argument `n_days` which is the option's time to expiration in days. The present value of a quantity of stock bought or sold on day $k$ is therefore calculated using the multiplicative factor $e^{-rk/252}$, since we need to convert to years.
- The function `garch_stock_paths` below generates a 2-D array each of whose rows is a simulated stock price path using the GARCH volatility model. The function is set up to return both the stock price paths and the volatilities $\sigma_t$ at each step, since we will need these volatilities later when simulating hedging along each path.
- `garch_MC_call` calls `garch_stock_paths` and then simulates the delta-hedged call option along those price paths. We assume the hedges are equally spaced, so the interval between hedges is `n_days//n_hedges` days (floor division). At each hedging interval, we need the Delta of the call option. We use the Black-Scholes explicit formula for Delta, where the volatility $\sigma$ is the GARCH-modeled volatility for that day (which `garch_stock_paths` returned along with the stock price for that day). In reality, an investor hedging a call option position would be more likely to use the *implied* volatility determined from option prices observed in the market that day, which is known as the "practitioner Black-Scholes" approach.
- The function `garch_MC_call` returns not only the simulated value of the call option portfolio with the required number of hedges (assuming the option is based on $100$ times the value of the index), but also two other numbers which function as measures of risk: the **Value at Risk** (VaR) and the **expected shortfall** (ES). We will explain the meaning of these numbers below, but for now we simply remark that we are using a $97.5\%$ threshold for VaR.
- Currently (6/20/2025, the day the GARCH calibration above took place), the S&P 500 index is about 5960 and the yield on 10-year U.S. Treasury notes is 4.39%. Therefore we use $S = 5960$ for the spot price and $r = 0.0439$ for the risk-free rate in the calculations at the end of the code block below. The S&P index is not a dollar amount, but using $100$ times the numerical value of the index to settle option contracts in cash at expiration is commonly done, e.g., by the Cboe Options Exchange with their SPX options.

```python
# Explicit Black-Scholes formula for delta

def bs_call_delta(S_0, K, sigma, T = 1, r = 0):
    
    d_1 = (np.log(S_0/K) + (r+.5*sigma**2)*T)/(sigma*np.sqrt(T))
    return norm.cdf(d_1)
```

```python
# Simulate stock price paths using GARCH(1,1) volatility process-- store and return volatility paths as well

def garch_stock_paths(S_0, r, omega, alpha, beta, n_sims, n_days): 

    """
    Parameters:
    S_0 (float): spot price
    r (float): risk-free interest rate
    omega, alpha, beta (floats): GARCH(1,1) parameters
    n_sims (int): number of simulations
    n_days (int): time to expiration in days
    
    Return:
    list of two np.arrays
    first array is a (n_sims, n_days + 1) grid of simulated stock price paths (one row for each simulation)
    second array (same size) records the GARCH-simulated volatilities (sigmas) for each day
    """
    
    dt = 1/252 # basic time increment = 1 day
    
    paths = np.zeros((n_sims, n_days + 1))
    vols = np.zeros((n_sims, n_days + 1)) # "sigmas" in GARCH
    volvols = np.zeros((n_sims, n_days + 1)) # "zetas" in GARCH
    # generate two grids of independent standard normals
    garch_noise_grid = np.random.normal(0, 1, size = (n_sims, n_days + 1)) 
    mc_noise_grid = np.random.normal(0, 1, size = (n_sims, n_days + 1))
    
    paths[:, 0] = S_0 # all paths start with the same price
    volvols[:, 0] = omega # zeta_0
    vols[:, 0] = volvols[:, 0]*garch_noise_grid[:, 0] # sigma_0 = zeta_0 times a standard normal

    # implement the GARCH recursion column-by-column
    # "full truncation" for negative volatilities (if sigma is negative, which makes no sense, replace it with |sigma|)
    for i in range(n_days):
        volvols[:, i+1] = np.sqrt(omega + alpha*vols[:, i]**2 + beta*vols[:, i]**2)
        vols[:, i+1] = volvols[:, i+1]*garch_noise_grid[:, i+1]
        paths[:, i+1] = paths[:, i]*np.exp((r-.5*vols[:, i])*dt + np.sqrt(np.abs(vols[:, i]*dt))*mc_noise_grid[:, i])    
        
    return [paths, vols]

def garch_MC_call(S_0, K, r, omega, alpha, beta, n_sims, n_days, n_hedges):

    """
    Parameters:
    S_0 (float): spot price
    K (float): strike price
    r (float): risk-free interest rate
    omega, alpha, beta (floats): GARCH(1,1) parameters
    n_sims (int): number of simulations
    n_days (int): time to expiration in days
    n_hedges (int): number of hedging periods (equally-spaced)
    
    Return:
    list of three floats
    first is the average value of the hedged call option (mean of the values under all n_sims simulations)
    second is the value at risk of the position (97.5% threshold)
    third is the expected shortfall 
    """
    
    hf = n_days//n_hedges # hedge frequency (number of days between hedges)
    T = n_days/252 # time to expiration in years
    
    [garch_paths, garch_vols] = garch_stock_paths(S_0, r, omega, alpha, beta, n_sims, n_days)

    path_end_points = garch_paths[:,-1] # spot price at expiration
    call_payouts = np.maximum(path_end_points - K,0)*np.exp(-r*T) # present value of call's value at expiration
    
    # Simulate stock profits (PnL of the position in stock used to delta-hedge) at each interval

    stock_profits = []

    for i in range(1,n_hedges):
        stock_start = garch_paths[:, (i-1)*hf]
        stock_end = garch_paths[:, i*hf]
        deltas = bs_call_delta(stock_start, K, garch_vols[:, i*hf], (n_days - i*hf)/252, r) # use GARCH sigmas in Black-Scholes formula
        
        stock_profit = ((stock_end - stock_start*np.exp(r*(hf/252)))*deltas*np.exp(-i*(hf/252)*r)) # hf/252 is the time in years between hedges
        stock_profits.append(stock_profit)

    total_stock_profit = np.sum(stock_profits, axis = 0)

    profits_hedged = call_payouts - total_stock_profit # final value of portfolio consisting of option + stock used to hedge

    mean_profit = np.mean(profits_hedged) # average value of portfolios over all simulations
    value_at_risk = -np.quantile(profits_hedged, 0.025) # 97.5 VaR (losses are quoted as positive VaR)
    expected_shortfall = -np.mean(profits_hedged[profits_hedged < -value_at_risk]) 
    
    return [100*mean_profit, 100*value_at_risk, 100*expected_shortfall] # options are on 100 times the index (100 "shares")

print(garch_MC_call(5960, 5960, 0.0439, 0.00000331, 0.08228, 0.8228, 10000, 126, 13)) # at-the-money option expiring in six months (252/2 days)
                                                                                     # hedged approximately every other week (52/(2*2) hedges)
```

In the example at the end of the code block above, we simulate the value of an at-the-money option on the S&P 500 (one hundred times the index's numerical value) expiring in six months and Delta-hedged every other week. The result we obtain is that the average value is \\$16732.31, with a Value at Risk (**VaR**, at the $97.5$% threshold) of \\$8012.80 and an expected shortfall of \\$12281.21. These last two numbers need additional explanation.

The motivation behind the concept of VaR is the desire to be able to make statements of the form "we are $X$% sure we will not lose more than $N$ dollars in the next $k$ days." The VaR in this sentence is $N$ dollars. We know how many days we are simulating, but how do we choose the value of $X$? For a VaR statement to be meaningful, it needs to be either explicitly stated or clear from context what $X$ is. Common choices are $X = 95$, $97.5$, and $99$. We have chosen to use $X = 97.5$ in this mini-project because the Basel Committee on Bank Supervision (in the *Fundamental Review of the Trading Book*) recommended basing capital requirements for banks on expected shortfall with a $97.5$% confidence level. 

Calculating VaR in the above simulation is easy; we just find the $0.025$ quantile of the array of simulated portfolio values. In the specific example we computed at the end, we ran $10000$ simulations. If these simulations' outcomes were ranked from first to $10000$th in increasing order, the $0.025$ quantile is the $251$st on the list, since $2.5$% of the simulations had an outcome as bad as or worse than this one. In most applications, VaR is a negative number (representing a loss) but is quoted as a positive number, the minus sign being understood. In our example, this means that $2.5$% of the time, we can expect to lose \\$8128.32 or more. Rephrased in the above form, "we are $97.5$% sure we will not lose more than \\$8128.32 at the end of six months."

VaR is appealing because it condenses risk into a single number (dollar amount). However, it tells us nothing about *how bad* things can get at the left tail of the distribution. The related concept of **expected shortfall** is used for this purpose. The expected shortfall is the average loss *conditioned on* the loss being worse than the VaR. In the case of a $97.5$% confidence VaR, expected shortfall is the mean of the lowest $2.5$\% of outcomes; in our example with $10000$ simulations, the expected shortfall, which was \\$12903.15, is the average of the $250$ worst outcomes (like the VaR, the minus sign is understood).

Below, we plot the results of $10000$ simulations where the number of hedges is taken to be one, two, six (monthly hedging), twelve (biweekly hedging), and $126$ (daily hedging). The VaR in each case is represented by a vertical red line.

```python
# code above copied but modified to return all simulation outcomes, not just the average, so we can make a histogram

def garch_MC_call_complete(S_0, K, r, omega, alpha, beta, n_sims, n_days, n_hedges):

    """
    Parameters:
    S_0 (float): spot price
    K (float): strike price
    r (float): risk-free interest rate
    omega, alpha, beta (floats): GARCH(1,1) parameters
    n_sims (int): number of simulations
    n_days (int): time to expiration in days
    n_hedges (int): number of hedging periods (equally-spaced)
    
    Return:
    list of three floats
    first is the average value of the hedged call option (mean of the values under all n_sims simulations)
    second is the value at risk of the position (97.5% threshold)
    third is the expected shortfall 
    """
    
    hf = n_days//n_hedges # hedge frequency (number of days between hedges)
    T = n_days/252 # time to expiration in years
    
    [garch_paths, garch_vols] = garch_stock_paths(S_0, r, omega, alpha, beta, n_sims, n_days)

    path_end_points = garch_paths[:,-1] # spot price at expiration
    call_payouts = np.maximum(path_end_points - K,0)*np.exp(-r*T) # present value of call's value at expiration
    
    # Simulate stock profits (PnL of the position in stock used to delta-hedge) at each interval

    stock_profits = []

    for i in range(1,n_hedges):
        stock_start = garch_paths[:, (i-1)*hf]
        stock_end = garch_paths[:, i*hf]
        deltas = bs_call_delta(stock_start, K, garch_vols[:, i*hf], (n_days - i*hf)/252, r) # use GARCH sigmas in Black-Scholes formula
        
        stock_profit = ((stock_end - stock_start*np.exp(r*(hf/252)))*deltas*np.exp(-i*(hf/252)*r)) # hf/252 is the time in years between hedges
        stock_profits.append(stock_profit)

    total_stock_profit = np.sum(stock_profits, axis = 0)

    profits_hedged = 100*(call_payouts - total_stock_profit) # final value of portfolio consisting of option + stock used to hedge

    return profits_hedged

hedging_numbers = [1, 2, 6, 12, 126]

for n_hedges in hedging_numbers:

    garch_profits_hedged = garch_MC_call_complete(5960, 5960, 0.0439, 0.00000331, 0.08228, 0.8228, 10000, 126, 13)

    plt.figure(figsize=(12,8))

    plt.hist(garch_profits_hedged, bins = 50, alpha = .4, color = 'black', label = 'Simulated Values')

    plt.axvline(np.mean(garch_profits_hedged), label = f'Simulated Mean: ${np.mean(garch_profits_hedged):.2f}', color = 'blue')
    
    plt.axvline(np.quantile(garch_profits_hedged, 0.025), label = f'Value at Risk: ${-np.quantile(garch_profits_hedged, 0.025):.2f}', color = 'red')

    plt.title(f'Distribution of simulated GARCH(1,1) values with 10000 simulations and {n_hedges} control variate(s)', size = 12, wrap = True)

    plt.legend()
    
    plt.show()
```
